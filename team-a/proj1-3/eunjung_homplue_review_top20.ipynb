```python
import requests
import pandas as pd

# 기본 URL 및 파라미터 설정
base_url = "https://mfront.homeplus.co.kr/review/getItemReview.json"

# 아이템 번호와 페이지 수 설정
orders = [
    {"itemNo": "068836278", "total_pages": 39},
    {"itemNo": "070606076", "total_pages": 11},
    {"itemNo": "069283215", "total_pages": 56},
    {"itemNo": "069739616", "total_pages": 20},
    {"itemNo": "070653670", "total_pages": 7},
    {"itemNo": "070171145", "total_pages": 28}
]


# 각 order에 대해 리뷰 데이터를 수집하고 데이터프레임으로 저장
for idx, order in enumerate(orders, start=1):
    item_no = order["itemNo"]
    total_pages = order["total_pages"]

    params = {
        "groupItem": "N",
        "isAttached": "false",
        "itemNo": item_no,
        "myStore": "N",
        "page": 1,  # 시작 페이지
        "perPage": 30,  # 페이지당 항목 수
        "reviewFilter": "",
        "sort": "reg",
        "storeKind": "NOR",
        "storeType": "HYPER",
        "type": "A"
    }

    # 전체 데이터를 저장할 리스트
    all_reviews = []

    # 페이지별로 데이터 수집
    for page in range(1, total_pages + 1):
        params["page"] = page  # 현재 페이지 설정
        response = requests.get(base_url, params=params)

        # 응답 상태 코드 확인
        if response.status_code == 200:
            data = response.json()
            reviews = data.get("data", [])
            all_reviews.extend(reviews)  # 전체 리뷰 리스트에 추가
            print(f"아이템 {item_no} - 페이지 {page} 수집 완료. 총 {len(reviews)}개의 리뷰 수집됨.")
        else:
            print(f"아이템 {item_no} - 페이지 {page} 요청 실패: 상태 코드 {response.status_code}")
            break

    # 판다스 데이터프레임으로 변환
    df = pd.DataFrame(all_reviews)

    # 데이터프레임을 각 파일로 저장 (순차적으로 df1, df2, ...로 저장)
    df.to_csv(f"homeplus_reviews_order{idx}.csv", index=False, encoding="utf-8-sig")
    print(f"아이템 {item_no}의 데이터 저장 완료. 'homeplus_reviews_order{idx}.csv'로 저장됨.")

print("모든 데이터 수집 및 저장 완료.")
```

```python
# CSV 파일 리스트
csv_files = [
    "homeplus_reviews_order1.csv",
    "homeplus_reviews_order2.csv",
    "homeplus_reviews_order3.csv",
    "homeplus_reviews_order4.csv",
    "homeplus_reviews_order5.csv",
    "homeplus_reviews_order6.csv"
]

# 모든 CSV 파일을 읽어들여 하나의 데이터프레임으로 합치기
combined_df = pd.concat([pd.read_csv(file) for file in csv_files], ignore_index=True)

# 합친 데이터프레임을 새로운 CSV 파일로 저장
combined_df.to_csv("homeplus_reviews_combined3.csv", index=False, encoding="utf-8-sig")

print("모든 리뷰 데이터가 하나의 CSV 파일로 합쳐졌습니다.")
```

```python
import duckdb

# DuckDB 데이터베이스 파일 경로
db_path = "homeplus_review_top20.db"

# DuckDB 데이터베이스 연결
conn = duckdb.connect(database=db_path)

# combined_df를 'reviews'라는 테이블로 DuckDB에 저장
conn.execute("CREATE TABLE IF NOT EXISTS reviews AS SELECT * FROM combined_df")

# 테이블 확인
result = conn.execute("SELECT COUNT(*) FROM reviews").fetchall()
print(f"테이블 'reviews'에 저장된 데이터 수: {result[0][0]}")

# DuckDB 연결 종료
conn.close()

print(f"데이터베이스 파일이 저장되었습니다: {db_path}")
```
