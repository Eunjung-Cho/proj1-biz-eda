```python
import re
```

```python

# 제거할 불용어 리스트
stopwords = ['밀키트', '홈밀', '요리재료', 'G', 'KG', 'x', '개', '인분', '직화', 'g']

# 정규화로 단어 정제하는 함수
def clean_and_tokenize(text):
    # 특수 문자와 숫자 제거
    text = re.sub(r'[^\w\s]', '', text)  # 특수 문자 제거
    text = re.sub(r'\d+', '', text)      # 숫자 제거
    # 소문자 변환 및 단어 분리
    words = text.lower().split()
    # 불용어 제거
    words = [word for word in words if word not in stopwords]
    # 단어들을 공백으로 연결하여 반환
    return ' '.join(words)
```

```python

# 상품명2 컬럼 추가 및 값 채우기
df_hp['상품명2'] = df_hp['상품명'].apply(clean_and_tokenize)
```

```python
from difflib import SequenceMatcher
```

```python


def preprocess_text(text):
    """
    텍스트 전처리 함수
    - 공백 제거
    - 특수문자 제거
    - 모두 소문자로 변환
    """
    # 공백 제거
    text = ''.join(text.split())
    # 특수문자 제거
    text = re.sub(r'[^\w\s]', '', text)
    # 소문자 변환
    return text.lower()

def calculate_similarity(text1, text2):
    """
    두 문자열 간의 유사도를 계산
    """
    return SequenceMatcher(None, text1, text2).ratio()

def find_best_match(source_text, target_list, threshold=0.7):
    """
    주어진 텍스트와 가장 유사한 항목을 target_list에서 찾기
    threshold: 최소 유사도 기준
    """
    source_text = preprocess_text(source_text)
    best_match = None
    best_ratio = 0
    
    for target in target_list:
        processed_target = preprocess_text(target)
        ratio = calculate_similarity(source_text, processed_target)
        
        if ratio > best_ratio:
            best_ratio = ratio
            best_match = target
    
    if best_ratio >= threshold:
        return best_match, best_ratio
    return None, best_ratio

# 매칭 실행 예시
def match_dataframes(df3, df_hp):
    matches = []
    for ckg_nm in df3['CKG_NM']:
        match, ratio = find_best_match(ckg_nm, df_hp['상품명2'].unique())
        matches.append({
            'CKG_NM': ckg_nm,
            'matched_name': match,
            'similarity': ratio
        })
    return pd.DataFrame(matches)
```
