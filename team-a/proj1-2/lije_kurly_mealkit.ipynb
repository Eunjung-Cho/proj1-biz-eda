```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python

# 웹드라이버 설정
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# URL 리스트
urls = [
    "https://www.kurly.com/categories/912011?page=1&per_page=96&sorted_type=1",
    "https://www.kurly.com/categories/912011?page=2&per_page=96&sorted_type=1",
    "https://www.kurly.com/categories/912011?page=3&per_page=96&sorted_type=1",
    "https://www.kurly.com/categories/912011?page=4&per_page=96&sorted_type=1"
]

# 상품 정보를 저장할 리스트
products = []
```

```python
for url in urls:
    driver.get(url)
    time.sleep(5)  # 페이지 로딩을 기다림

    # 상품명 추출
    product_names = driver.find_elements(By.CSS_SELECTOR, "#container > div.css-pzlq5x.ev4ofwi1 > div.css-1d3w5wq.ef36txc6 > div.css-11kh0cw.ef36txc5 > a > div.css-1kpzrna.e1c07x486 > span.css-1dry2r1.e1c07x485")
    
    # 가격 추출
    product_prices = driver.find_elements(By.CSS_SELECTOR, "#container > div.css-pzlq5x.ev4ofwi1 > div.css-1d3w5wq.ef36txc6 > div.css-11kh0cw.ef36txc5 > a > div.css-1kpzrna.e1c07x486 > div.e1c07x484.discount-price.css-bwx2kz.ei5rudb2 > div.discount > span.sales-price.css-105zgxh.ei5rudb1 > span.price-number")
    
    # 리뷰 수 추출
    product_reviews = driver.find_elements(By.CSS_SELECTOR, "#container > div.css-pzlq5x.ev4ofwi1 > div.css-1d3w5wq.ef36txc6 > div.css-11kh0cw.ef36txc5 > a > div.css-1kpzrna.e1c07x486 > div.review-count.css-xezqwk.e1c07x482 > span.review-number.css-l610kl.e1c07x480")

    # 상품 정보를 리스트에 저장
    for name, price, review in zip(product_names, product_prices, product_reviews):
        products.append({
            'name': name.text,
            'price': price.text,
            'review': review.text
        })

# 웹드라이버 종료
driver.quit()

# 결과 출력
for product in products:
    print(f"상품명: {product['name']}, 가격: {product['price']}, 리뷰 수: {product['review']}")
```

# 데이터 정제

- 리뷰수를 명확히 표시해주지 않아 가장 많은 리뷰갯수표기법인 9,999+ 인 제품 찾기

```python
# 리뷰 수가 '9,999+'인 제품 개수 세기
count = 0
for product in products:
    if product['review'] == '9,999+':
        count += 1

print(f"리뷰 수가 '9,999+'인 제품의 개수: {count}")
```

```python
# 가격을 숫자로 변환하는 함수
def price_to_number(price):
    return int(price.replace(',', ''))  # 쉼표 제거 후 정수로 변환

# 리뷰 수가 '9,999+'인 제품과 그렇지 않은 제품 분류
high_review_products = []
low_review_products = []

for product in products:
    if product['review'] == '9,999+':
        high_review_products.append(product)
    else:
        low_review_products.append(product)

# 각 그룹의 가격 평균 계산
def calculate_average_price(product_list):
    if not product_list:  # 리스트가 비어있는 경우
        return 0
    total_price = sum(price_to_number(product['price']) for product in product_list)
    return total_price / len(product_list)

high_review_avg_price = calculate_average_price(high_review_products)
low_review_avg_price = calculate_average_price(low_review_products)

# 결과 출력
print(f"리뷰 수가 '9,999+'인 제품의 평균 가격: {high_review_avg_price:.2f}원")
print(f"리뷰 수가 '9,999+'가 아닌 제품의 평균 가격: {low_review_avg_price:.2f}원")
```

### 제품의 가격대는 판매량의 큰 영향을 주지 않음.

```python
# 회사명 추출 함수
def extract_company_name(product_name):
    return product_name.split(']')[0][1:]  # '['와 ']' 사이의 회사명 추출

# 리뷰 수를 숫자로 변환하는 함수
def review_to_number(review):
    if review.endswith('+'):
        return int(review[:-1].replace(',', ''))  # '999+' 또는 '1,000+' 형식을 숫자로 변환
    return int(review.replace(',', ''))  # 숫자만 있는 경우

# 회사별 리뷰 수 데이터 저장
company_reviews = {}

for product in products:
    company = extract_company_name(product['name'])
    review = review_to_number(product['review'])
    
    if company not in company_reviews:
        company_reviews[company] = {'total_reviews': 0, 'count': 0}
    
    company_reviews[company]['total_reviews'] += review
    company_reviews[company]['count'] += 1

# 회사별 리뷰 수 평균 계산
company_avg_reviews = {}
for company, data in company_reviews.items():
    avg_review = data['total_reviews'] / data['count']
    company_avg_reviews[company] = avg_review

# 회사별 평균 리뷰 수를 정렬 (내림차순)
sorted_company_avg_reviews = sorted(company_avg_reviews.items(), key=lambda x: x[1], reverse=True)

# 결과 출력
for company, avg_review in sorted_company_avg_reviews:
    print(f"회사명: {company}, 평균 리뷰 수: {avg_review:.2f}")
```

```python
# 가격을 숫자로 변환하는 함수
def price_to_number(price):
    return int(price.replace(',', ''))  # 쉼표 제거 후 정수로 변환

# 리뷰 수를 숫자로 변환하는 함수
def review_to_number(review):
    if review.endswith('+'):
        return int(review[:-1].replace(',', ''))  # '999+' 또는 '1,000+' 형식을 숫자로 변환
    return int(review.replace(',', ''))  # 숫자만 있는 경우

# 판매 총액 계산 및 상품 정보 업데이트
for product in products:
    price = price_to_number(product['price'])
    review = review_to_number(product['review'])
    product['total_sales'] = price * review  # 판매 총액 계산

# 판매 총액을 기준으로 상위 5개 상품 추출
top_5_products = sorted(products, key=lambda x: x['total_sales'], reverse=True)[:5]

# 결과 출력
print("가장 판매 총액이 높은 상위 5개 상품:")
for product in top_5_products:
    print(f"상품명: {product['name']}, 가격: {product['price']}, 리뷰 수: {product['review']}, 판매 총액: {product['total_sales']:,}원")
```

```python
import pandas as pd
from scipy.stats import pearsonr  
```

```python
# 가격과 리뷰 수를 숫자로 변환
def price_to_number(price):
    return int(price.replace(',', ''))  # 쉼표 제거 후 정수로 변환

def review_to_number(review):
    if review.endswith('+'):
        return int(review[:-1].replace(',', ''))  # '999+' 또는 '1,000+' 형식을 숫자로 변환
    return int(review.replace(',', ''))  # 숫자만 있는 경우

# 데이터프레임 생성
data = {
    'price': [price_to_number(p['price']) for p in products],
    'review': [review_to_number(p['review']) for p in products]
}
df = pd.DataFrame(data)

# 상관계수 계산
correlation, _ = pearsonr(df['price'], df['review'])

# 결과 출력
print(f"가격과 리뷰 수의 상관계수: {correlation:.4f}")
```

### 약한 양의 상관관계 가격이 높을수록 리뷰수 증가하는 경향을 보이나 상관계수가 0에 가깝기 때문에, 가격과 리뷰 수는 서로 독립적이라고 볼 수 있으며 리뷰 수는 가격 외에 다른 요인(예: 제품 품질, 브랜드 인지도, 마케팅 등)에 의해 더 큰 영향을 받을 가능성이 높습니다.
