```python
import requests
import json
import pandas as pd
import duckdb

# Define the URL
url = "https://mfront.homeplus.co.kr/review/getItemReview.json"

# Function to fetch reviews for a specific page
def fetch_reviews(page):
    params = {
        "groupItem": "N",
        "isAttached": "false",
        "itemNo": "070278031",
        "myStore": "N",
        "page": page,
        "perPage": 30,
        "reviewFilter": "",
        "sort": "reg",
        "storeKind": "NOR",
        "storeType": "HYPER",
        "type": "A"
    }

    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        if data["returnStatus"] == 200:
            return data.get("data", [])
    return []

try:
    # Collect all reviews by iterating through pages
    all_reviews = []
    page = 1

    while True:
        print(f"Fetching page {page}...")
        reviews = fetch_reviews(page)
        if not reviews:  # Stop if no more reviews are found
            break
        all_reviews.extend(reviews)
        page += 1

    # Prepare a list to store review data
    review_list = []
    for review in all_reviews:
        review_data = {
            "Review ID": review.get("reviewNo"),
            "User ID": review.get("userId"),
            "User Name": review.get("userNm"),
            "Item Name": review.get("itemNm"),
            "Grade": review.get("grade"),
            "Review Contents": review.get("contents"),
            "Review Date": review.get("regDt"),
            "Repurchase": "Yes" if review.get("repurchaseYn") == "Y" else "No",
            "Recommendation": "Yes" if review.get("recommendYn") == "Y" else "No",
            "Image URLs": [img.get("imgUrl") for img in review.get("imageList", [])]
        }
        review_list.append(review_data)

    # Create a DataFrame from the review list
    df = pd.DataFrame(review_list)

    # Save the DataFrame to DuckDB
    conn = duckdb.connect(database='reviews.duckdb', read_only=False)
    conn.execute("CREATE TABLE IF NOT EXISTS reviews AS SELECT * FROM df")
    conn.execute("INSERT INTO reviews SELECT * FROM df")
    conn.close()

    print("All reviews have been saved to DuckDB.")

except Exception as e:
    print("An error occurred:", e)
```

```python
# prompt: df['Grade']['Review Contents'] 을 바탕으로 5개 주제로 토픽 모델링하고 주요 토픽별 키워드를 추출할것

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Assuming 'df' is your DataFrame and 'Review Contents' column exists
# If not, replace 'df' and 'Review Contents' with your actual DataFrame and column name

# Preprocess the text data (optional but recommended)
# This example removes punctuation, converts to lowercase, and tokenizes
import re
def preprocess_text(text):
    text = re.sub(r'[^\w\s]', '', str(text))  # Remove punctuation
    text = text.lower()  # Convert to lowercase
    return text

df['Processed Review'] = df['Review Contents'].apply(preprocess_text)


# TF-IDF vectorization
vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, max_features=1000, stop_words='english')  # Adjust parameters as needed
tfidf = vectorizer.fit_transform(df['Processed Review'].astype(str))

# LDA topic modeling
num_topics = 5
lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda.fit(tfidf)

# Get top keywords for each topic
feature_names = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(lda.components_):
    top_keywords_idx = topic.argsort()[:-10 - 1:-1]  # Get indices of top 10 keywords
    top_keywords = [feature_names[i] for i in top_keywords_idx]
    print(f"Topic {topic_idx + 1}: {', '.join(top_keywords)}")
```

```python
# prompt: 4개의 군집으로 위 리뷰를 군집화 하고 군집별로 TF-IDF 값의 합계를 구해 판다스코드로 높은 순으로 10개씩 출력하는 코드작성

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Assuming 'df' and 'Processed Review' column exist as in the previous code
# ... (previous code to create df and 'Processed Review' column)

# TF-IDF Vectorization
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['Processed Review'].astype(str))

# KMeans Clustering
num_clusters = 4
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(tfidf_matrix)
df['Cluster'] = kmeans.labels_

# Calculate TF-IDF sum for each cluster and top 10 keywords
for cluster_num in range(num_clusters):
    cluster_data = df[df['Cluster'] == cluster_num]
    cluster_tfidf = tfidf_matrix[cluster_data.index]

    tfidf_sum = cluster_tfidf.sum(axis=0)
    feature_names = vectorizer.get_feature_names_out()
    tfidf_df = pd.DataFrame({'keyword': feature_names, 'tfidf_sum': tfidf_sum.tolist()[0]})
    tfidf_df = tfidf_df.sort_values(by='tfidf_sum', ascending=False).head(10)
    print(f"Top 10 TF-IDF Keywords for Cluster {cluster_num + 1}:")
    print(tfidf_df)
    print("-" * 50)
```

```python
# 머메이드 COLAP에는 따로 설치해야함
sequenceDiagram
    participant 사용자 as 사용자
    participant 스크립트 as 파이썬 스크립트
    participant 웹서버 as 웹/API 서버
    participant 데이터베이스 as 데이터베이스

    사용자->>스크립트: 데이터 크롤링 시작
    스크립트->>웹서버: HTTP 요청 전송 (GET/POST)
    웹서버-->>스크립트: 응답 반환 (JSON/HTML)

    스크립트->>스크립트: 응답 데이터 파싱
    alt 응답이 JSON 형식일 경우
        스크립트->>스크립트: JSON 데이터 추출
    else 응답이 HTML 형식일 경우
        스크립트->>스크립트: BeautifulSoup으로 HTML 파싱
    end

    스크립트->>스크립트: 데이터 정리 및 가공
    스크립트->>데이터베이스: 데이터베이스 연결
    데이터베이스-->>스크립트: 연결 성공

    스크립트->>데이터베이스: 테이블 생성 (필요한 경우)
    스크립트->>데이터베이스: 데이터 삽입
    데이터베이스-->>스크립트: 데이터 삽입 완료

    스크립트->>웹서버: 다음 페이지 요청 (있을 경우)
    웹서버-->>스크립트: 다음 페이지 데이터 반환
    스크립트->>스크립트: 더 이상 데이터가 없을 때까지 반복

    스크립트-->>사용자: 데이터 크롤링 완료
    사용자->>데이터베이스: 데이터 확인 요청
    데이터베이스-->>사용자: 저장된 데이터 반환
```

```python
import requests
import pandas as pd

# 요청 URL과 헤더 정보
url = "https://m.ctrip.com/restapi/soa2/18109/json/getAttractionList"
headers = {
    "Content-Type": "application/json",
    "Referer": "https://m.ctrip.com/",
    "User-Agent": "Mozilla/5.0"
}

# 요청 페이로드 정보
payload = {
    "head": {
        "cid": "09031058213352606128",
        "ctok": "",
        "cver": "1.0",
        "lang": "01",
        "sid": "8888",
        "syscode": "999",
        "auth": "",
        "xsid": "",
        "extension": []
    },
    "scene": "online",
    "districtId": 1446512,
    "index": 4,
    "sortType": 1,
    "count": 10,
    "filter": {
        "filterItems": []
    },
    "returnModuleType": "product"
}

# 데이터 요청
response = requests.post(url, headers=headers, json=payload)
response.raise_for_status()
data = response.json()

# 데이터에서 관광지 정보 추출
attraction_list = data.get("attractionList", [])

# 관광지 정보를 데이터프레임으로 변환
attractions = []
for attraction in attraction_list:
    card = attraction.get("card", {})
    attractions.append({
        "Name": card.get("poiName"),
        "Zone": card.get("zoneName"),
        "CommentCount": card.get("commentCount"),
        "CommentScore": card.get("commentScore"),
        "District": card.get("districtName"),
        "ImageURL": card.get("coverImageUrl"),
        "Distance": card.get("distanceStr"),
        "Tags": ", ".join(card.get("tagNameList", [])),
        "OpenStatus": card.get("openStatus"),
        "ShortFeatures": ", ".join(card.get("shortFeatures", [])),
        "Latitude": card.get("coordinate", {}).get("latitude"),
        "Longitude": card.get("coordinate", {}).get("longitude"),
        "DetailURL": card.get("detailUrlInfo", {}).get("url")
    })

# 판다스 데이터프레임 생성
df = pd.DataFrame(attractions)

# 데이터프레임 저장 또는 출력
print(df.head())
# df.to_csv("attractions.csv", index=False, encoding="utf-8-sig")
```
