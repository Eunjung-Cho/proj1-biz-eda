```python
import requests
import pandas as pd

# 기본 URL 및 파라미터 설정
base_url = "https://mfront.homeplus.co.kr/review/getItemReview.json"
params = {
    "groupItem": "N",
    "isAttached": "false",
    "itemNo": "070278031",
    "myStore": "N",
    "page": 1,  # 시작 페이지
    "perPage": 30,  # 페이지당 항목 수
    "reviewFilter": "",
    "sort": "reg",
    "storeKind": "NOR",
    "storeType": "HYPER",
    "type": "A"
}

# 전체 데이터를 저장할 리스트
all_reviews = []

# 총 페이지 수 (예: 191페이지)
total_pages = 191

# 페이지별로 데이터 수집
for page in range(1, total_pages + 1):
    params["page"] = page  # 현재 페이지 설정
    response = requests.get(base_url, params=params)

    # 응답 상태 코드 확인
    if response.status_code == 200:
        data = response.json()
        reviews = data.get("data", [])
        all_reviews.extend(reviews)  # 전체 리뷰 리스트에 추가
        print(f"페이지 {page} 수집 완료. 총 {len(reviews)}개의 리뷰 수집됨.")
    else:
        print(f"페이지 {page} 요청 실패: 상태 코드 {response.status_code}")
        break

# 판다스 데이터프레임으로 변환
df = pd.DataFrame(all_reviews)

# 데이터프레임 저장 (CSV 파일로 저장)
df.to_csv("homeplus_reviews.csv", index=False, encoding="utf-8-sig")
print("모든 데이터 수집 및 저장 완료.")
```

```python
# prompt: df_all_reviews[[contents','grade' ]]을 바탕으로 5개의 주제로 토픽 모델링을 하고 주요 토픽별 키워드를 시각화 함

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import matplotlib.pyplot as plt
import koreanize_matplotlib

# 데이터 로드 (homeplus_reviews.csv 파일이 존재해야 함)
df_all_reviews = pd.read_csv("homeplus_reviews.csv")

# 'contents'와 'grade' 컬럼으로 토픽 모델링 수행
documents = df_all_reviews['contents'].astype(str)  # Ensure 'contents' is string type

# TF-IDF 벡터화
vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english') # Adjust parameters if needed
tfidf = vectorizer.fit_transform(documents)

# LDA 모델 학습 (5개의 주제)
num_topics = 5
lda_model = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda_model.fit(tfidf)

# 주요 토픽별 키워드 출력
feature_names = vectorizer.get_feature_names_out()

for topic_idx, topic in enumerate(lda_model.components_):
  top_features_ind = topic.argsort()[:-10 - 1:-1]
  top_features = [feature_names[i] for i in top_features_ind]
  print(f"Topic #{topic_idx + 1}: {', '.join(top_features)}")


# 토픽 분포 시각화
plt.figure(figsize=(10, 6))
plt.bar(range(num_topics), lda_model.components_.sum(axis=1))  # 각 토픽의 단어 분포 합계 시각화
plt.xlabel('토픽')
plt.ylabel('단어 분포')
plt.title('주제별 단어 분포')
plt.xticks(range(num_topics), [f'Topic {i+1}' for i in range(num_topics)]) #토픽 번호 추가
plt.show()


# 각 문서의 토픽 분포 확인 (선택 사항)
topic_distribution = lda_model.transform(tfidf)

# 예시: 상위 3개 토픽 출력
for i, topic_probs in enumerate(topic_distribution[:5]): # 상위 5개 리뷰 출력
    top_topics = topic_probs.argsort()[-3:][::-1]  # 상위 3개 토픽
    print(f"Review {i+1}: Top Topics - {[f'Topic {topic + 1}' for topic in top_topics]}, Probabilities - {[topic_probs[topic] for topic in top_topics]}")
```

```python
# prompt: 4개 군집으로 위 리뷰를 군집화 하고 군집별로 TF-IDF 값의 합계를 구해 판다스코드로 높은 순으로 10개씩 가중치값과 함께 출력할 것
# numpy 사용

import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# 데이터 로드 (homeplus_reviews.csv 파일이 존재해야 함)
df_all_reviews = pd.read_csv("homeplus_reviews.csv")

# 'contents' 컬럼 전처리 (결측치 처리 등)
documents = df_all_reviews['contents'].fillna('').astype(str)

# TF-IDF 벡터화
vectorizer = TfidfVectorizer(max_df=0.95, min_df=2)
tfidf_matrix = vectorizer.fit_transform(documents)

# KMeans 군집화 (4개 군집)
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(tfidf_matrix)

# 군집 레이블 추가
df_all_reviews['cluster'] = kmeans.labels_

# 군집별 TF-IDF 합계 계산 및 상위 10개 출력
for cluster_num in range(4):
    cluster_data = df_all_reviews[df_all_reviews['cluster'] == cluster_num]
    cluster_tfidf = tfidf_matrix[cluster_data.index]
    tfidf_sum = np.array(cluster_tfidf.sum(axis=0)).flatten()

    # TF-IDF 값과 단어 매핑
    tfidf_df = pd.DataFrame({'word': vectorizer.get_feature_names_out(), 'tfidf': tfidf_sum})

    # TF-IDF 값 기준으로 내림차순 정렬
    tfidf_df = tfidf_df.sort_values(by='tfidf', ascending=False)

    # 상위 10개 출력
    print(f"Cluster {cluster_num + 1} Top 10 TF-IDF:")
    print(tfidf_df.head(10))
    print("-" * 30)
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

# 샘플 HTML을 직접 입력하거나 웹에서 가져올 수 있습니다.
url = "https://www.ssfshop.com/public/search/popup/searchGoodList"
params = {
    "serviceType": "SRC",
    "keyword": "잠옷",
    "orderView": "SORT_WEIGHT",
    "pageNo": 2,
    "_csrf": "2c1b91ff-1717-4747-826c-4832f4fb2158",
    "tab": "GOD",
    "schType": "ALL",
}

# HTTP GET 요청
response = requests.get(url, params=params)

if response.status_code == 200:
    # BeautifulSoup으로 HTML 파싱
    soup = BeautifulSoup(response.text, 'html.parser')

    # 상품 정보 추출
    items = soup.find_all('div', class_='god-info')

    data = []
    for item in items:
        # 브랜드명
        brand_tag = item.find('span', class_='brand')
        brand = brand_tag.get_text(strip=True) if brand_tag else None

        # 상품명
        name_tag = item.find('span', class_='name')
        name = name_tag.get_text(strip=True) if name_tag else None

        # 원가
        original_price_tag = item.find('del')
        original_price = original_price_tag.get_text(strip=True).replace('원가', '') if original_price_tag else None

        # 할인율
        discount_rate_tag = item.find('em', class_='sale')
        discount_rate = discount_rate_tag.get_text(strip=True) if discount_rate_tag else None

        # 판매가
        sale_price_tag = item.find('span', class_='price').find('span', class_='wa_hidden', text='판매가')
        sale_price = sale_price_tag.find_next_sibling(text=True).strip() if sale_price_tag else None

        # 평점
        rating_tag = item.find('span', class_='point')
        rating = float(rating_tag.find('em').get_text(strip=True)) if rating_tag else None

        # 리뷰 수
        review_count_tag = item.find('span', class_='review')
        review_count = int(review_count_tag.find('em').get_text(strip=True)) if review_count_tag else None

        # 찜 수
        likes_tag = item.find('span', class_='heart')
        likes = int(likes_tag.find('em').get_text(strip=True)) if likes_tag else None

        # 데이터 추가
        data.append({
            "brand": brand,
            "name": name,
            "original_price": original_price,
            "discount_rate": discount_rate,
            "sale_price": sale_price,
            "rating": rating,
            "review_count": review_count,
            "likes": likes
        })

    # DataFrame 생성
    df = pd.DataFrame(data)

    # 결과 확인
    print(df.head())
else:
    print(f"HTTP 요청 실패. 상태 코드: {response.status_code}")
```

```python
import requests
import pandas as pd

# 요청 URL 및 파라미터
url = "https://www.lotteon.com/search/search/search.ecn"
params = {
    "u2": "120",
    "u3": "60",
    "render": "qapi",
    "platform": "pc",
    "collection_id": "9",
    "q": "버버리 가방",
    "mallId": "1",
    "u39": "0",
}

# 요청 헤더 설정
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
}

# HTTP GET 요청
response = requests.get(url, params=params, headers=headers)

# 응답 데이터 확인 및 처리
if response.status_code == 200:
    response_data = response.json()
    item_list = response_data.get("itemList", [])

    # 데이터 추출
    data = []
    for item in item_list:
        product_name = item.get("productName", "")
        brand_name = item.get("brandName", "")
        original_price = next((price["num"] for price in item.get("priceInfo", []) if price["type"] == "original"), None)
        final_price = next((price["num"] for price in item.get("priceInfo", []) if price["type"] == "final"), None)
        discount_rate = next((price["num"] for price in item.get("priceInfo", []) if price["type"] == "discount"), None)
        product_link = item.get("productLink", "")
        product_image = item.get("productImage", "")
        delivery_info = next((info["value"] for info in item.get("additionalInfo", []) if info["type"] == "fd"), "")

        # 데이터 리스트에 추가
        data.append({
            "Product Name": product_name,
            "Brand Name": brand_name,
            "Original Price": original_price,
            "Final Price": final_price,
            "Discount Rate (%)": discount_rate,
            "Product Link": product_link,
            "Product Image": product_image,
            "Delivery Info": delivery_info,
        })

    # Pandas DataFrame 생성
    df = pd.DataFrame(data)

    # 결과 출력
    print(df.head())

    # 데이터 저장
    df.to_csv("lotteon_products.csv", index=False, encoding="utf-8-sig")
    print("크롤링 결과가 lotteon_products.csv 파일에 저장되었습니다.")
else:
    print(f"HTTP 요청 실패. 상태 코드: {response.status_code}")
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# 요청 URL 및 파라미터
base_url = "https://www.lotteon.com/search/search/search.ecn"
params = {
    "u2": "120",
    "u3": "60",
    "render": "qapi",
    "platform": "pc",
    "collection_id": "9",
    "q": "버버리 가방",
    "mallId": "1",
    "u39": "0",
}

# 요청 헤더 설정
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
}

# GET 요청으로 메인 데이터 크롤링
response = requests.get(base_url, params=params, headers=headers)

if response.status_code == 200:
    response_data = response.json()
    item_list = response_data.get("itemList", [])

    # 메인 페이지 데이터 수집
    main_data = []
    for item in item_list:
        product_name = item.get("productName", "")
        brand_name = item.get("brandName", "")
        original_price = next((price["num"] for price in item.get("priceInfo", []) if price["type"] == "original"), None)
        final_price = next((price["num"] for price in item.get("priceInfo", []) if price["type"] == "final"), None)
        discount_rate = next((price["num"] for price in item.get("priceInfo", []) if price["type"] == "discount"), None)
        product_link = item.get("productLink", "")
        product_image = item.get("productImage", "")
        delivery_info = next((info["value"] for info in item.get("additionalInfo", []) if info["type"] == "fd"), "")

        # 메인 페이지 데이터 저장
        main_data.append({
            "Product Name": product_name,
            "Brand Name": brand_name,
            "Original Price": original_price,
            "Final Price": final_price,
            "Discount Rate (%)": discount_rate,
            "Product Link": product_link,
            "Product Image": product_image,
            "Delivery Info": delivery_info,
        })

    # DataFrame 생성
    main_df = pd.DataFrame(main_data)
    print("메인 데이터 수집 완료")

    # 상세 페이지 순회
    detailed_data = []
    for product in main_data:
        detail_url = product["Product Link"]
        print(f"상세 페이지 요청: {detail_url}")

        detail_response = requests.get(detail_url, headers=headers)
        if detail_response.status_code == 200:
            soup = BeautifulSoup(detail_response.text, 'html.parser')

            # 상세 페이지에서 추가 데이터 추출 (예: 상세 정보, 리뷰 등)
            description = soup.find('div', {'class': 'product-description'})  # 상세 설명
            description_text = description.get_text(strip=True) if description else "No description"

            detailed_data.append({
                "Product Name": product["Product Name"],
                "Detail Description": description_text,
            })
        else:
            print(f"상세 페이지 요청 실패: {detail_url}")

    # 상세 데이터 DataFrame 생성
    detail_df = pd.DataFrame(detailed_data)
    print("상세 페이지 데이터 수집 완료")

    # 결과 통합
    result_df = pd.merge(main_df, detail_df, on="Product Name", how="left")

    # 저장
    result_df.to_csv("lotteon_products_with_details.csv", index=False, encoding="utf-8-sig")
    print("전체 데이터 저장 완료")
else:
    print(f"메인 페이지 요청 실패. 상태 코드: {response.status_code}")
```
