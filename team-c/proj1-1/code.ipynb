```python
import matplotlib.pyplot as plt
import pandas as pd

# 데이터 준비
data = {
    '연도': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    '중국': [2000000, 2200000, 2500000, 2400000, 2300000, 2100000, 500000, 600000, 1700000, 1800000],
    '일본': [500000, 480000, 470000, 460000, 450000, 440000, 100000, 120000, 430000, 450000],
    '미국': [100000, 110000, 120000, 130000, 140000, 150000, 50000, 60000, 110000, 120000],
    '기타': [200000, 210000, 220000, 230000, 240000, 250000, 70000, 80000, 240000, 250000]
}

df = pd.DataFrame(data)
df['총계'] = df[['중국', '일본', '미국', '기타']].sum(axis=1)

# 시각화
plt.figure(figsize=(12, 6))
for country in ['중국', '일본', '미국', '기타']:
    plt.plot(df['연도'], df[country], marker='o', label=country)

plt.plot(df['연도'], df['총계'], marker='o', linestyle='--', color='black', label='총계')

plt.title('연도별 제주도 방문 외국인 관광객 수')
plt.xlabel('연도')
plt.ylabel('방문객 수')
plt.legend()
plt.grid(True)
plt.xticks(df['연도'])
plt.tight_layout()
plt.show()
```
