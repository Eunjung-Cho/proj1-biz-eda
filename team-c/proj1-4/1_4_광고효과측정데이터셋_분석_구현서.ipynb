# 추천 광고 캠페인 분석 지표 및 기법 (우선순위 순)

## 핵심 성과 지표 (KPI)

1. **ROAS (광고 투자 수익률)**
   - 계산: 매출액 ÷ 광고_지출액
   - 중요도: 최우선 - 마케팅 투자의 직접적 수익성 평가
   - 활용: 채널별, 광고형식별, 타겟팅별 투자 효율성 비교

2. **전환율 (Conversion Rate)**
   - 계산: 전환 수 ÷ 웹사이트_방문_수
   - 중요도: 매우 높음 - 방문자의 구매 전환 효율성 측정
   - 활용: 광고 메시지와 랜딩페이지의 효과 평가

3. **CPA (전환당 비용)**
   - 계산: 광고_지출액 ÷ 전환 수
   - 중요도: 높음 - 고객 획득 비용 측정
   - 활용: 비용 효율적인 채널 및 캠페인 식별

4. **CTR (클릭률)**
   - 계산: (광고_클릭_횟수 ÷ 광고_노출_횟수) × 100%
   - 중요도: 높음 - 광고 메시지와 크리에이티브의 효과 측정
   - 활용: 광고 형식 및 메시지 최적화

5. **광고_태도_변화**
   - 범위: -1(부정적) ~ 1(긍정적)
   - 중요도: 중간-높음 - 브랜드 인식 변화 측정
   - 활용: 장기적 브랜드 가치 구축 전략 수립

## 추천 분석 기법 (우선순위 순)

1. **채널별 효율성 비교 분석**
   - 방법: 각 채널(TV, 온라인, SNS)의 ROAS, CPA, CTR 비교
   - 목적: 최적 채널 믹스 결정
   - 결과물: 채널별 성과 대시보드, 채널 효율성 순위

2. **세그먼트 성과 분석**
   - 방법: 인구통계(연령, 성별, 직업, 지역)와 관심사별 반응 비교
   - 목적: 고성과 고객 세그먼트 발견
   - 결과물: 세그먼트별 성과 히트맵, 세그먼트 프로파일링

3. **타겟팅 정확도 분석**
   - 방법: 광고_타겟팅과 실제 반응 사용자 특성 매칭 비교
   - 목적: 타겟팅 전략 효과 검증 및 최적화
   - 결과물: 타겟팅 정확도 지수, 타겟 그룹별 성과 차트

4. **광고 형식 효과 분석**
   - 방법: 배너, 동영상, 텍스트, 이미지별 성과 비교
   - 목적: 최적 광고 형식 결정
   - 결과물: 형식별 성과 비교 차트, 채널-형식 조합 성과 매트릭스

5. **전환 퍼널 분석**
   - 방법: 노출→클릭→방문→페이지뷰→전환 단계별 이탈률 분석
   - 목적: 전환 과정의 병목 지점 식별
   - 결과물: 퍼널 시각화, 단계별 이탈률 차트

6. **예산 최적화 모델링**
   - 방법: 지출-성과 관계 회귀 분석 및 최적화 알고리즘
   - 목적: 최적 예산 할당 결정
   - 결과물: 한계 수익 곡선, 최적 예산 배분 시나리오

7. **시계열 효과 분석**
   - 방법: 광고 기간에 따른 CTR, 전환율, 회상률 변화 추적
   - 목적: 최적 캠페인 기간 및 타이밍 결정
   - 결과물: 시간에 따른 성과 트렌드 그래프, 계절성 패턴 차트

8. **머신러닝 기반 예측 모델링**
   - 방법: 로지스틱 회귀, 랜덤 포레스트 등을 활용한 클릭/전환 예측
   - 목적: 고전환 가능성 사용자 식별 및 타겟팅 최적화
   - 결과물: 예측 모델, 변수 중요도 분석, 성향 점수(Propensity Score)

9. **고객 생애 가치(CLV) 분석**
   - 방법: 구매_이력과 현재 구매 기반 미래 가치 예측
   - 목적: 장기적 고객 관계 가치 평가
   - 결과물: CLV 세그먼트 맵, 고가치 고객 프로필

10. **기여도 모델링(Attribution Modeling)**
    - 방법: 다중 채널 접점의 전환 기여도 할당
    - 목적: 각 채널의 실제 기여도 평가
    - 결과물: 채널별 기여도 차트, 터치포인트 가중치 모델

이러한 분석은 순차적으로 진행하면서 각 단계의 인사이트를 다음 분석에 적용하는 것이 효과적입니다. 첫 5개 분석은 기본적인 성과 측정과 최적화에 필수적이며, 나머지 분석들은 더 깊은 인사이트와 전략적 의사결정을 위한 고급 분석에 해당합니다.

#**CTR이란?**

CTR은 광고가 노출된 횟수 중에서 실제로 클릭된 비율을 나타냅니다. 즉, 사용자들이 광고를 보고 얼마나 관심을 보였는지를 측정하는 지표입니다.

## **계산방식**

`CTR = (클릭 수 / 노출 수) × 100%`

````python
import pandas as pd
import numpy as np

# 데이터프레임이 df로 로드되었다고 가정
# CTR 계산: (클릭 수 / 노출 수) * 100
df['CTR'] = (df['광고_클릭_횟수'] / df['광고_노출_횟수']) * 100

# CTR 소수점 둘째 자리까지 표시
df['CTR'] = df['CTR'].round(2)

# 결과 확인 (처음 5개 행)
print(df[['광고_채널', '광고_형식', '광고_노출_횟수', '광고_클릭_횟수', 'CTR']].head())

# 채널별 평균 CTR 계산
channel_ctr = df.groupby('광고_채널')['CTR'].mean().reset_index()
print("\n채널별 평균 CTR:")
print(channel_ctr)

# 광고 형식별 평균 CTR 계산
format_ctr = df.groupby('광고_형식')['CTR'].mean().reset_index()
print("\n광고 형식별 평균 CTR:")
print(format_ctr)

# 채널 및 형식별 CTR 계산
channel_format_ctr = df.groupby(['광고_채널', '광고_형식'])['CTR'].mean().reset_index()
print("\n채널 및 형식별 평균 CTR:")
print(channel_format_ctr)
````

# ROAS (Return On Ad Spend) 설명

## ROAS란?
ROAS(Return On Ad Spend)는 광고 지출 수익률을 의미하며, 광고에 투자한 비용 대비 창출된 매출의 비율을 나타냅니다. 마케팅 효율성을 측정하는 핵심 지표로, 투자한 광고비 대비 얼마나 많은 수익을 얻었는지 직접적으로 보여줍니다.

## ROAS 계산 방식
```
ROAS = (광고로 발생한 매출액 / 광고 지출액) × 100%
```

예를 들어, 광고에 1,000,000원을 지출하고 5,000,000원의 매출이 발생했다면:
ROAS = (5,000,000 / 1,000,000) × 100% = 500%

## ROAS의 중요성
1. **마케팅 효율성 측정**: 어떤 광고 캠페인이 가장 효율적인지 파악할 수 있습니다
2. **예산 최적화**: 높은 ROAS를 보이는 채널이나 캠페인에 예산을 집중할 수 있습니다
3. **비즈니스 의사결정**: 마케팅 전략 수립과 광고 지출 규모 결정에 중요한 기준이 됩니다
4. **성과 벤치마킹**: 업계 평균과 비교하여 광고 성과를 평가할 수 있습니다

## ROAS vs ROI
- ROAS는 매출/비용의 비율을 측정하며, 주로 마케팅 팀에서 사용합니다
- ROI는 (이익-비용)/비용의 비율을 측정하며, 순이익을 고려합니다


````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 데이터가 이미 로드되었다고 가정
# df = pd.read_csv('광고_데이터.csv')

# ROAS 계산 (퍼센트로 표현)
df['ROAS'] = (df['매출액'] / df['광고_지출액']) * 100

# 기본 통계 확인
roas_stats = df['ROAS'].describe()
print("ROAS 기본 통계:")
print(roas_stats)

# 이상치 처리 (선택사항 - 매우 극단적인 값들은 분석에 왜곡을 줄 수 있음)
# 상위 99 백분위수 이상은 이상치로 간주
upper_limit = np.percentile(df['ROAS'], 99)
df_filtered = df[df['ROAS'] <= upper_limit]

# 채널별 평균 ROAS 계산
channel_roas = df.groupby('광고_채널')['ROAS'].agg(['mean', 'median', 'std', 'count'])
print("\n채널별 ROAS 통계:")
print(channel_roas)

# 광고 형식별 평균 ROAS 계산
format_roas = df.groupby('광고_형식')['ROAS'].agg(['mean', 'median', 'std', 'count'])
print("\n광고 형식별 ROAS 통계:")
print(format_roas)

# 채널과 형식 조합별 ROAS
combined_roas = df.groupby(['광고_채널', '광고_형식'])['ROAS'].mean().reset_index()
print("\n채널-형식 조합별 평균 ROAS:")
print(combined_roas)

# 타겟팅별 ROAS 분석
# 타겟팅 문자열에서 연령대 추출 (예: "20대, 남성, 패션" -> "20대")
df['타겟_연령대'] = df['광고_타겟팅'].str.extract(r'(\d+대)')
df['타겟_성별'] = df['광고_타겟팅'].str.extract(r'[0-9]+대, (남성|여성)')
df['타겟_관심사'] = df['광고_타겟팅'].str.extract(r'[0-9]+대, (?:남성|여성), (.+)')

# 타겟 연령대별 ROAS
age_roas = df.groupby('타겟_연령대')['ROAS'].mean().reset_index()
print("\n타겟 연령대별 평균 ROAS:")
print(age_roas)

# 타겟 성별 ROAS
gender_roas = df.groupby('타겟_성별')['ROAS'].mean().reset_index()
print("\n타겟 성별별 평균 ROAS:")
print(gender_roas)

# 타겟 관심사별 ROAS
interest_roas = df.groupby('타겟_관심사')['ROAS'].mean().reset_index()
print("\n타겟 관심사별 평균 ROAS:")
print(interest_roas)

# 시각화: 채널별 ROAS 박스플롯
plt.figure(figsize=(12, 6))
sns.boxplot(x='광고_채널', y='ROAS', data=df_filtered)
plt.title('채널별 ROAS 분포')
plt.xlabel('광고 채널')
plt.ylabel('ROAS (%)')
plt.grid(True, alpha=0.3)
plt.axhline(y=100, color='r', linestyle='--', alpha=0.7)  # 손익분기점 라인 (ROAS 100%)
plt.show()

# 시각화: 형식별 ROAS 바차트
plt.figure(figsize=(12, 6))
sns.barplot(x='광고_형식', y='ROAS', data=df, ci=None)
plt.title('광고 형식별 평균 ROAS')
plt.xlabel('광고 형식')
plt.ylabel('평균 ROAS (%)')
plt.grid(True, alpha=0.3)
plt.axhline(y=100, color='r', linestyle='--', alpha=0.7)  # 손익분기점 라인
plt.show()

# ROAS와 다른 지표들과의 상관관계 분석
correlation_cols = ['ROAS', 'CTR', '전환율', '광고_회상률', '광고_태도_변화', '웹사이트_방문_수', '페이지뷰']
correlation = df[correlation_cols].corr()
print("\nROAS와 주요 지표 간 상관관계:")
print(correlation['ROAS'])

# 시각화: 상관관계 히트맵
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('ROAS와 주요 성과 지표 간 상관관계')
plt.tight_layout()
plt.show()

# 추가 분석: ROAS 구간별 캠페인 수 분포
roas_bins = [0, 50, 100, 200, 500, 1000, float('inf')]
roas_labels = ['매우 낮음 (<50%)', '낮음 (50-100%)', '적정 (100-200%)',
               '좋음 (200-500%)', '매우 좋음 (500-1000%)', '탁월함 (>1000%)']
df['ROAS_구간'] = pd.cut(df['ROAS'], bins=roas_bins, labels=roas_labels)

roas_distribution = df['ROAS_구간'].value_counts().sort_index()
print("\nROAS 구간별 캠페인 수:")
print(roas_distribution)

# 시각화: ROAS 구간별 분포
plt.figure(figsize=(12, 6))
sns.countplot(x='ROAS_구간', data=df, order=roas_labels)
plt.title('ROAS 구간별 캠페인 수 분포')
plt.xlabel('ROAS 구간')
plt.ylabel('캠페인 수')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3, axis='y')
plt.show()

# 추가 분석: ROAS와 광고 지출액의 관계 (효율적인 예산 사용 파악)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='광고_지출액', y='ROAS', data=df, alpha=0.5, hue='광고_채널')
plt.title('광고 지출액과 ROAS의 관계')
plt.xlabel('광고 지출액')
plt.ylabel('ROAS (%)')
plt.axhline(y=100, color='r', linestyle='--', alpha=0.7)  # 손익분기점 라인
plt.legend(title='광고 채널')
plt.grid(True, alpha=0.3)
plt.show()

# 마케팅 효율성 지표: ROAS가 100% 이상인 캠페인 비율
profitable_campaigns = (df['ROAS'] >= 100).mean() * 100
print(f"\n수익성 있는 캠페인 비율 (ROAS ≥ 100%): {profitable_campaigns:.2f}%")

# 채널별 수익성 있는 캠페인 비율
channel_profitability = df.groupby('광고_채널').apply(lambda x: (x['ROAS'] >= 100).mean() * 100)
print("\n채널별 수익성 있는 캠페인 비율:")
print(channel_profitability)
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import matplotlib.ticker as ticker

# 데이터가 이미 로드되었다고 가정
# df = pd.read_csv('광고_데이터.csv')

# 1. 기본 통계 및 분포 분석
attitude_stats = df['광고_태도_변화'].describe()
print("광고_태도_변화 기본 통계:")
print(attitude_stats)

# 분포 시각화
plt.figure(figsize=(10, 6))
sns.histplot(df['광고_태도_변화'], bins=50, kde=True)
plt.title('광고_태도_변화 분포')
plt.xlabel('광고_태도_변화 (-1 ~ 1)')
plt.ylabel('빈도')
plt.axvline(x=0, color='red', linestyle='--', alpha=0.7)
plt.grid(True, alpha=0.3)
plt.show()

# 2. 채널별 태도 변화 분석
channel_attitude = df.groupby('광고_채널')['광고_태도_변화'].agg(['mean', 'median', 'std', 'min', 'max']).reset_index()
print("\n채널별 광고_태도_변화 통계:")
print(channel_attitude)

# 채널별 태도 변화 분포
plt.figure(figsize=(12, 6))
sns.boxplot(x='광고_채널', y='광고_태도_변화', data=df)
plt.title('채널별 광고_태도_변화 분포')
plt.xlabel('광고 채널')
plt.ylabel('광고_태도_변화 (-1 ~ 1)')
plt.grid(True, alpha=0.3)
plt.show()

# 3. 광고 형식별 태도 변화 분석
format_attitude = df.groupby('광고_형식')['광고_태도_변화'].agg(['mean', 'median', 'std', 'min', 'max']).reset_index()
print("\n광고 형식별 광고_태도_변화 통계:")
print(format_attitude)

# 광고 형식별 태도 변화 시각화
plt.figure(figsize=(12, 6))
sns.barplot(x='광고_형식', y='광고_태도_변화', data=df, estimator=np.mean, errorbar=('ci', 95))
plt.title('광고 형식별 평균 광고_태도_변화')
plt.xlabel('광고 형식')
plt.ylabel('평균 광고_태도_변화')
plt.grid(True, alpha=0.3, axis='y')
plt.show()

# 4. 태도 변화와 주요 성과 지표 간 상관관계
# 필요한 지표들만 선택
performance_metrics = ['광고_태도_변화', 'CTR', '전환율', 'ROAS', '광고_회상률', '매출액', '페이지뷰']
correlation = df[performance_metrics].corr()
print("\n광고_태도_변화와 주요 성과 지표 간 상관관계:")
print(correlation['광고_태도_변화'])

# 상관관계 히트맵
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('광고_태도_변화와 주요 성과 지표 간 상관관계')
plt.tight_layout()
plt.show()

# 5. 연령별 태도 변화 분석
df['타겟_연령대'] = df['광고_타겟팅'].str.extract(r'(\d+대)')
age_attitude = df.groupby('타겟_연령대')['광고_태도_변화'].mean().reset_index()
print("\n타겟 연령대별 평균 광고_태도_변화:")
print(age_attitude)

# 연령별 태도 변화 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x='타겟_연령대', y='광고_태도_변화', data=age_attitude)
plt.title('타겟 연령대별 평균 광고_태도_변화')
plt.xlabel('타겟 연령대')
plt.ylabel('평균 광고_태도_변화')
plt.grid(True, alpha=0.3, axis='y')
plt.show()

# 6. 성별에 따른 태도 변화 분석
df['타겟_성별'] = df['광고_타겟팅'].str.extract(r'[0-9]+대, (남성|여성)')
gender_attitude = df.groupby('타겟_성별')['광고_태도_변화'].mean().reset_index()
print("\n타겟 성별별 평균 광고_태도_변화:")
print(gender_attitude)

# 성별 태도 변화 시각화
plt.figure(figsize=(8, 6))
sns.barplot(x='타겟_성별', y='광고_태도_변화', data=gender_attitude)
plt.title('타겟 성별별 평균 광고_태도_변화')
plt.xlabel('타겟 성별')
plt.ylabel('평균 광고_태도_변화')
plt.grid(True, alpha=0.3, axis='y')
plt.show()

# 7. 관심사별 태도 변화 분석
df['타겟_관심사'] = df['광고_타겟팅'].str.extract(r'[0-9]+대, (?:남성|여성), (.+)')
interest_attitude = df.groupby('타겟_관심사')['광고_태도_변화'].mean().reset_index()
print("\n타겟 관심사별 평균 광고_태도_변화:")
print(interest_attitude)

# 관심사별 태도 변화 시각화
plt.figure(figsize=(12, 6))
sns.barplot(x='타겟_관심사', y='광고_태도_변화', data=interest_attitude)
plt.title('타겟 관심사별 평균 광고_태도_변화')
plt.xlabel('타겟 관심사')
plt.ylabel('평균 광고_태도_변화')
plt.grid(True, alpha=0.3, axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 8. 태도 변화 수준별 캠페인 성과 비교
# 태도 변화를 구간으로 분류
attitude_bins = [-1.0, -0.5, -0.1, 0.1, 0.5, 1.0]
attitude_labels = ['매우 부정적', '부정적', '중립적', '긍정적', '매우 긍정적']
df['태도_변화_구간'] = pd.cut(df['광고_태도_변화'], bins=attitude_bins, labels=attitude_labels)

# 태도 변화 구간별 주요 지표 평균 비교
metrics_by_attitude = df.groupby('태도_변화_구간')[['CTR', '전환율', 'ROAS', '광고_회상률']].mean().reset_index()
print("\n태도 변화 구간별 주요 성과 지표:")
print(metrics_by_attitude)

# 태도 변화 구간별 성과 지표 시각화 (멀티바 차트)
metrics_list = ['CTR', '전환율', 'ROAS', '광고_회상률']
for metric in metrics_list:
    plt.figure(figsize=(10, 6))
    sns.barplot(x='태도_변화_구간', y=metric, data=metrics_by_attitude)
    plt.title(f'태도 변화 구간별 평균 {metric}')
    plt.xlabel('태도 변화 구간')
    plt.ylabel(f'평균 {metric}')
    plt.grid(True, alpha=0.3, axis='y')
    plt.show()

# 9. 광고 기간과 태도 변화의 관계 분석
# 광고 기간 계산 (일수)
df['광고_시작일'] = pd.to_datetime(df['광고_시작일'])
df['광고_종료일'] = pd.to_datetime(df['광고_종료일'])
df['광고_기간'] = (df['광고_종료일'] - df['광고_시작일']).dt.days + 1

# 광고 기간별 태도 변화 분석
period_groups = [1, 7, 14, 21, 30]
df['광고_기간_구간'] = pd.cut(df['광고_기간'], bins=[0] + period_groups + [float('inf')],
                        labels=['1일', '2-7일', '8-14일', '15-21일', '22-30일', '30일 초과'])

period_attitude = df.groupby('광고_기간_구간')['광고_태도_변화'].mean().reset_index()
print("\n광고 기간별 평균 광고_태도_변화:")
print(period_attitude)

# 광고 기간별 태도 변화 시각화
plt.figure(figsize=(12, 6))
sns.barplot(x='광고_기간_구간', y='광고_태도_변화', data=period_attitude)
plt.title('광고 기간별 평균 광고_태도_변화')
plt.xlabel('광고 기간')
plt.ylabel('평균 광고_태도_변화')
plt.grid(True, alpha=0.3, axis='y')
plt.show()

# 10. 채널-형식 조합별 태도 변화 히트맵
channel_format_attitude = df.groupby(['광고_채널', '광고_형식'])['광고_태도_변화'].mean().unstack()
print("\n채널-형식 조합별 평균 광고_태도_변화:")
print(channel_format_attitude)

plt.figure(figsize=(10, 8))
sns.heatmap(channel_format_attitude, annot=True, cmap='RdBu_r', center=0, fmt='.3f')
plt.title('채널-형식 조합별 평균 광고_태도_변화')
plt.tight_layout()
plt.show()

# 11. 태도 변화가 가장 높은/낮은 캠페인 특성 분석
top_positive = df.nlargest(10, '광고_태도_변화')[['광고_채널', '광고_형식', '광고_타겟팅', '광고_태도_변화']]
print("\n태도 변화가 가장 긍정적인 상위 10개 캠페인:")
print(top_positive)

top_negative = df.nsmallest(10, '광고_태도_변화')[['광고_채널', '광고_형식', '광고_타겟팅', '광고_태도_변화']]
print("\n태도 변화가 가장 부정적인 상위 10개 캠페인:")
print(top_negative)

# 12. 직업별 태도 변화 분석
job_attitude = df.groupby('직업')['광고_태도_변화'].mean().reset_index()
print("\n직업별 평균 광고_태도_변화:")
print(job_attitude)

# 직업별 태도 변화 시각화
plt.figure(figsize=(12, 6))
sns.barplot(x='직업', y='광고_태도_변화', data=job_attitude)
plt.title('직업별 평균 광고_태도_변화')
plt.xlabel('직업')
plt.ylabel('평균 광고_태도_변화')
plt.grid(True, alpha=0.3, axis='y')
plt.show()
````

3. **장기 캠페인의 긍정적 효과**: 30일 이상의 캠페인이 상대적으로 가장 긍정적인 태도 변화를 보이는 것은 지속적인 노출이 브랜드 인식 개선에 도움이 될 수 있음을 암시합니다.

4. **세그먼트별 미묘한 차이**: 40대, 여성, 음악 관심사, 자영업자 등이 상대적으로 더 긍정적인 태도 변화를 보이지만, 그 차이는 매우 작습니다.

5. **온라인 채널의 소폭 우위**: 온라인 채널이 유일하게 양의 태도 변화 평균을 보이며, 특히 온라인 배너와 텍스트 형식이 상대적으로 효과적입니다.

이 결과는 광고 태도 변화가 매우 미묘하고 복합적인 요인에 영향을 받으며, 단기적인 성과 지표와는 다른 메커니즘으로 작용함을 시사합니다. 장기적인 브랜드 구축 전략을 세울 때 이러한 미묘한 차이를 고려할 필요가 있습니다.

````python
# 전환수 계산 (방문자 수 × 전환율)
df['전환수'] = df['웹사이트_방문_수'] * df['전환율']

# CPA(Cost Per Acquisition) 계산
df['CPA'] = df['광고_지출액'] / df['전환수']
df['CPA'] = df['CPA'].replace([np.inf, -np.inf], np.nan)  # 전환수가 0인 경우 처리

# 채널별 주요 지표 계산
channel_metrics = df.groupby('광고_채널').agg({
    'ROAS': 'mean',
    'CTR': 'mean',
    'CPA': 'mean',
    '광고_지출액': 'sum',
    '매출액': 'sum',
    '전환수': 'sum',
    '웹사이트_방문_수': 'sum',
    '광고_노출_횟수': 'sum',
    '광고_클릭_횟수': 'sum'
}).reset_index()

# 총 ROI 계산
channel_metrics['총_ROI'] = (channel_metrics['매출액'] - channel_metrics['광고_지출액']) / channel_metrics['광고_지출액'] * 100

# 효율성 순위 계산
channel_metrics['ROAS_순위'] = channel_metrics['ROAS'].rank(ascending=False)
channel_metrics['CTR_순위'] = channel_metrics['CTR'].rank(ascending=False)
channel_metrics['CPA_순위'] = channel_metrics['CPA'].rank(ascending=True)  # CPA는 낮을수록 좋음
channel_metrics['종합_순위'] = (channel_metrics['ROAS_순위'] + channel_metrics['CTR_순위'] + channel_metrics['CPA_순위']) / 3
````

````python
# 1. 핵심 성과 지표 비교 (ROAS, CTR, CPA)
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# ROAS 비교
sns.barplot(x='광고_채널', y='ROAS', data=channel_metrics, ax=axes[0], palette='viridis')
axes[0].set_title('채널별 평균 ROAS (%)')
axes[0].set_ylabel('ROAS (%)')
axes[0].axhline(y=100, color='r', linestyle='--', alpha=0.7)  # 손익분기점 표시
axes[0].grid(True, alpha=0.3)

# CTR 비교
sns.barplot(x='광고_채널', y='CTR', data=channel_metrics, ax=axes[1], palette='viridis')
axes[1].set_title('채널별 평균 CTR (%)')
axes[1].set_ylabel('CTR (%)')
axes[1].grid(True, alpha=0.3)

# CPA 비교
sns.barplot(x='광고_채널', y='CPA', data=channel_metrics, ax=axes[2], palette='viridis')
axes[2].set_title('채널별 평균 CPA (₩)')
axes[2].set_ylabel('CPA (₩)')
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 2. 볼륨 지표 비교 (노출, 클릭, 방문, 전환)
fig, axes = plt.subplots(1, 4, figsize=(20, 6))

metrics = ['광고_노출_횟수', '광고_클릭_횟수', '웹사이트_방문_수', '전환수']
titles = ['총 노출 수 (백만)', '총 클릭 수 (천)', '총 방문자 수 (천)', '총 전환 수 (천)']
dividers = [1000000, 1000, 1000, 1000]  # 단위 조정

for i, (metric, title, divider) in enumerate(zip(metrics, titles, dividers)):
    # 값을 적절한 단위로 변환
    channel_metrics[f'{metric}_adjusted'] = channel_metrics[metric] / divider

    sns.barplot(x='광고_채널', y=f'{metric}_adjusted', data=channel_metrics, ax=axes[i], palette='viridis')
    axes[i].set_title(title)
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 3. 효율성 레이더 차트
categories = ['ROAS', 'CTR', '전환율', 'CPA 효율성']

# 데이터 정규화 (0-1 스케일)
def normalize(series):
    if series.name == 'CPA':  # CPA는 낮을수록 좋음
        return (series.max() - series) / (series.max() - series.min())
    else:  # 다른 지표는 높을수록 좋음
        return (series - series.min()) / (series.max() - series.min())

norm_data = pd.DataFrame({
    'ROAS': normalize(channel_metrics['ROAS']),
    'CTR': normalize(channel_metrics['CTR']),
    '전환율': normalize(df.groupby('광고_채널')['전환율'].mean().reset_index()['전환율']),
    'CPA 효율성': normalize(channel_metrics['CPA'])
})

# 레이더 차트 그리기
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, polar=True)

angles = np.linspace(0, 2*np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]  # 원을 닫기 위해 첫 각도 반복

for i, channel in enumerate(channel_metrics['광고_채널']):
    values = norm_data.iloc[i].tolist()
    values += values[:1]  # 원을 닫기 위해 첫 값 반복

    ax.plot(angles, values, linewidth=2, label=channel)
    ax.fill(angles, values, alpha=0.25)

ax.set_thetagrids(np.degrees(angles[:-1]), categories)
ax.set_ylim(0, 1)
ax.set_title('채널별 효율성 레이더 차트', size=15)
ax.grid(True)
ax.legend(loc='upper right')

plt.tight_layout()
plt.show()
````

````python
# 채널별 ROI 계산
channel_metrics['ROI_percent'] = channel_metrics['총_ROI'].round(2)

# 채널 효율성 순위표 생성
efficiency_rank = channel_metrics[['광고_채널', 'ROAS', 'CTR', 'CPA', 'ROI_percent', '종합_순위']]
efficiency_rank = efficiency_rank.sort_values('종합_순위')

print("채널 효율성 순위:")
print(efficiency_rank)

# 채널 믹스 예산 배분 계산
total_budget = 100  # 총 예산을 100%로 가정
efficiency_score = 1 / efficiency_rank['종합_순위']  # 종합 순위의 역수
total_efficiency = efficiency_score.sum()
budget_allocation = (efficiency_score / total_efficiency * total_budget).round(2)

channel_recommendation = pd.DataFrame({
    '광고_채널': efficiency_rank['광고_채널'],
    '효율성_점수': efficiency_score.values,
    '권장_예산_비율': budget_allocation.values
})

print("\n채널 믹스 권장 예산 배분:")
print(channel_recommendation)
````
