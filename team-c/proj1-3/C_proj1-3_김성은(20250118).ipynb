```python
# prompt: 다음 데이터를 수치형으로 변환항 것 문자 데이터는 무시하고 결측치 처리할 것
# likes,
# original_price, sale_price 는 원가69,000원 => 69000 형식으로 문자를 제거하고 수치형으로 변환
# , sale_rate 는 할인율10% => 10 으로 문자를 제거하고 수치형으로 변환

import pandas as pd
import glob
import re

# 확장자가 csv인 파일 하나를 찾습니다.
csv_files = glob.glob('*.csv')

if csv_files:
  # 첫 번째 csv 파일을 읽어옵니다.
  df = pd.read_csv(csv_files[0])

  # 수치형으로 변환할 열 지정
  numeric_cols = ['likes', 'original_price', 'sale_price', 'sale_rate']

  # 각 열에 대해 전처리 및 변환 수행
  for col in numeric_cols:
    if col in df.columns:
        # 문자 제거 및 숫자 추출
        df[col] = df[col].astype(str).apply(lambda x: re.sub(r'[^\d]', '', x))

        # 빈 문자열을 NaN으로 변환 후 숫자형으로 변환, 에러 발생 시 NaN으로 처리
        df[col] = pd.to_numeric(df[col], errors='coerce')

  print(df.head())
  print(df.info())

else:
  print("No csv files found in the current directory.")
```

```python
# prompt: df 변수의 값을 정규화하고 새로운 변수에 할당해서 전체 박스플롯을 시각화

import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# 정규화할 열 선택 (수치형 열만 선택)
numeric_cols = df.select_dtypes(include=['number']).columns

# MinMaxScaler 객체 생성
scaler = MinMaxScaler()

# 선택한 열을 정규화하여 새로운 데이터프레임 생성
df_normalized = df.copy()
df_normalized[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# 전체 열에 대한 박스플롯 시각화
plt.figure(figsize=(12, 6))
df_normalized.boxplot()
plt.title("Boxplot of Normalized Data")
plt.xticks(rotation=45, ha='right')  # x축 레이블 회전
plt.tight_layout()
plt.show()
```
