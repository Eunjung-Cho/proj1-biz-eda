프롬프트: df["event_params"][0] 의 내부가 다음과 같습니다. 다음의 key 값을 컬럼명으로 value를 값으로 데이터프레임에 적용하는 코드 작성

```python
import pandas as pd
import json
from IPython.display import display

def parse_event_params(row):
    try:
        # JSON 문자열을 파이썬 딕셔너리로 변환
        params = json.loads(row)

        # 결과를 저장할 딕셔너리 초기화
        result = {}

        # event_params 리스트를 순회하며 key-value 쌍 추출
        for param in params['event_params']:
            key = param['key']
            # value 딕셔너리에서 None이 아닌 첫 번째 값을 찾음
            value_dict = param['value']
            value = next((v for v in value_dict.values() if v is not None), None)
            result[key] = value

        return pd.Series(result)
    except:
        return pd.Series()

# event_params 컬럼을 파싱하여 새로운 데이터프레임 생성
parsed_params = df['event_params'].apply(parse_event_params)

# 기존 데이터프레임과 병합
result_df = pd.concat([df.drop('event_params', axis=1), parsed_params], axis=1)

# 새로 추가된 컬럼 확인
new_columns = parsed_params.columns.tolist()
print("새로 추가된 컬럼들:")
print(new_columns)

# 결과 데이터프레임의 처음 5행 출력
display(result_df.head())

# 새로 추가된 컬럼들의 기초 통계
display(result_df[new_columns].describe())
```

```python
import pandas as pd
import json
from IPython.display import display

def is_json_column(series):
    # 첫 번째 non-null 값을 확인
    first_value = series.dropna().iloc[0] if not series.dropna().empty else None
    if isinstance(first_value, str):
        return first_value.strip().startswith('{')
    return False

def parse_json_value(value):
    if pd.isna(value):
        return {}
    try:
        if isinstance(value, str):
            parsed = json.loads(value)
            return flatten_dict(parsed)
        return {}
    except:
        return {}

def flatten_dict(d, parent_key='', sep='_'):
    items = []
    for k, v in d.items():
        new_key = f"{parent_key}{sep}{k}" if parent_key else k

        if isinstance(v, dict):
            items.extend(flatten_dict(v, new_key, sep=sep).items())
        elif isinstance(v, list):
            # 리스트인 경우 각 항목을 파싱
            for i, item in enumerate(v):
                if isinstance(item, dict):
                    items.extend(flatten_dict(item, f"{new_key}_{i}", sep=sep).items())
                else:
                    items.append((f"{new_key}_{i}", item))
        else:
            items.append((new_key, v))
    return dict(items)

# 원본 데이터프레임 복사
df_processed = df.copy()

# JSON 형식의 컬럼 찾기
json_columns = [col for col in df.columns if is_json_column(df[col])]
print(f"파싱할 JSON 컬럼들: {json_columns}")

# 각 JSON 컬럼 파싱
parsed_dataframes = []
columns_to_drop = []

for col in json_columns:
    print(f"\n{col} 컬럼 파싱 중...")
    parsed_series = df[col].apply(parse_json_value)
    parsed_df = pd.DataFrame.from_records(parsed_series.tolist())

    if not parsed_df.empty:
        # 컬럼명에 원본 컬럼명 추가
        parsed_df.columns = [f"{col}_{subcol}" for subcol in parsed_df.columns]
        parsed_dataframes.append(parsed_df)
        columns_to_drop.append(col)

# 모든 파싱된 데이터프레임 병합
non_json_df = df_processed.drop(columns=columns_to_drop)
result_dfs = [non_json_df] + parsed_dataframes
final_df = pd.concat(result_dfs, axis=1)

# 결과 출력
print("\n최종 데이터프레임 정보:")
print(f"원본 컬럼 수: {len(df.columns)}")
print(f"파싱 후 컬럼 수: {len(final_df.columns)}")
print("\n새로 생성된 컬럼들:")
new_columns = set(final_df.columns) - set(df.columns)
print(sorted(list(new_columns)))

# 결과 데이터프레임의 처음 5행 출력
display(final_df.head())

# 기초 통계 정보 출력
display(final_df.describe())
```
