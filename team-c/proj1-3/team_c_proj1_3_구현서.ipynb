```python
# prompt: 판다스로 같은 경로에 있는 csv 찾아서 데이터프레임 로드

import pandas as pd
import glob
import os

def load_csv_files_in_directory(directory):
  """
  Loads all CSV files in a given directory into a single Pandas DataFrame.

  Args:
      directory: The path to the directory containing the CSV files.

  Returns:
      A Pandas DataFrame containing the combined data from all CSV files, or None if no CSV files are found.
  """
  csv_files = glob.glob(os.path.join(directory, '*.csv'))  # Find all .csv files in the directory
  if not csv_files:
    print("No CSV files found in the specified directory.")
    return None

  all_dataframes = []
  for file in csv_files:
    try:
      df = pd.read_csv(file)  # Read each CSV file into a DataFrame
      all_dataframes.append(df)
    except pd.errors.EmptyDataError:
      print(f"Warning: Skipping empty file {file}")
    except pd.errors.ParserError:
      print(f"Warning: Skipping file with parsing error: {file}")

  if not all_dataframes:
    print("No valid CSV files found after processing.")
    return None

  combined_df = pd.concat(all_dataframes, ignore_index=True)  # Concatenate all DataFrames
  return combined_df

# Example usage:
current_directory = os.getcwd() # Get current working directory
combined_dataframe = load_csv_files_in_directory(current_directory)

if combined_dataframe is not None:
  print(combined_dataframe.head()) # Display the first few rows of the combined DataFrame
  print(combined_dataframe.info()) # Display DataFrame info
```

```python
# prompt: Combined Document에서 html 제거

import re

def remove_html_tags(text):
  """Removes HTML tags from a string."""
  clean = re.compile('<.*?>')
  return re.sub(clean, '', text)

if 'Combined Document' in df.columns:
    df['Combined Document'] = df['Combined Document'].apply(remove_html_tags)
    # Recalculate the length after removing HTML tags
    df['Combined Document Length'] = df['Combined Document'].str.len()

    # Visualize the distribution of document lengths
    plt.figure(figsize=(10, 6))
    plt.hist(df['Combined Document Length'], bins=50, color='skyblue', edgecolor='black')
    plt.xlabel('Document Length')
    plt.ylabel('Frequency')
    plt.title('Distribution of Combined Document Lengths (HTML tags removed)')
    plt.grid(axis='y', alpha=0.75)
    plt.show()

    # Print some statistics
    print(df['Combined Document Length'].describe())
else:
    print("Error: 'Combined Document' column not found in the DataFrame.")
```

```python
# prompt: Combined Document를 5개의 주제로 토픽모델링을 하고 각 토픽별 주요키워드 20개씩 가중치까지 보여줘

!pip install pyLDAvis
!pip install gensim

import pandas as pd
import numpy as np
from gensim import corpora, models
import pyLDAvis.gensim_models

# Assuming 'df' is your DataFrame with the 'Combined Document' column
# ... (your existing code to load and preprocess data) ...

# Prepare the text data for topic modeling
documents = df['Combined Document'].dropna().astype(str).tolist() # Handle potential NaN values

# Tokenize the documents (you might need more sophisticated tokenization)
texts = [[word for word in document.lower().split()] for document in documents]

# Create a dictionary from the tokenized documents
dictionary = corpora.Dictionary(texts)

# Create a corpus from the dictionary
corpus = [dictionary.doc2bow(text) for text in texts]

# Perform LDA topic modeling
num_topics = 5  # Set the number of topics
lda_model = models.LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics, passes=15) # Increased passes for better results

# Print top keywords for each topic
for topic_id in range(num_topics):
    top_keywords = lda_model.show_topic(topic_id, topn=20)
    print(f"Top 20 keywords for Topic #{topic_id + 1}:")
    for keyword, weight in top_keywords:
        print(f"  - {keyword}: {weight}")
    print("\n")

# Visualize the topics using pyLDAvis
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary)
vis
```

```python
# prompt: df['Combined Document'] 를 TF-IDF로 5개의 주제로 NMF로 토픽모델링을 하고 각 토픽별 주요 키워드와 가중치 20개씩 추출

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF

# Assuming 'df' is your DataFrame with the 'Combined Document' column
# ... (your existing code to load and preprocess data) ...

# Prepare the text data for topic modeling
documents = df['Combined Document'].dropna().astype(str).tolist()

# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')  # Adjust parameters as needed
tfidf = vectorizer.fit_transform(documents)

# Apply Non-negative Matrix Factorization (NMF)
num_topics = 5
nmf_model = NMF(n_components=num_topics, random_state=42)
nmf_model.fit(tfidf)

# Get feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Print top 20 keywords for each topic
for topic_idx, topic in enumerate(nmf_model.components_):
    top_keywords_indices = topic.argsort()[:-21:-1]  # Get indices of top 20 keywords
    top_keywords = [(feature_names[i], topic[i]) for i in top_keywords_indices]
    print(f"Top 20 keywords for Topic #{topic_idx + 1}:")
    for keyword, weight in top_keywords:
        print(f"  - {keyword}: {weight}")
    print("\n")
```

```python
# prompt: 유사도 분석을 하고 heatmap으로 상위 100개 문서만 시각화

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'tfidf' is your TF-IDF matrix and 'documents' is your list of documents
# ... (your existing code to load and preprocess data and calculate tfidf) ...

# Calculate cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(tfidf)

# Get top 100 documents based on a criterion (e.g., document length or another metric)
# Replace 'Combined Document Length' with the appropriate column name if needed.
top_100_indices = df.nlargest(100, 'Combined Document Length').index

# Extract the similarity scores for the top 100 documents
top_100_similarity = similarity_matrix[top_100_indices, :][:, top_100_indices]

# Visualize the heatmap
plt.figure(figsize=(12, 10))  # Adjust figure size as needed
sns.heatmap(top_100_similarity, cmap='viridis', xticklabels=False, yticklabels=False)
plt.title('Cosine Similarity Heatmap (Top 100 Documents)')
plt.show()
```

```python
# prompt: 위에서 구한 유사도를 바탕으로 랜덤하게 5새 문서를 가져와서 각 문서와 유사도가 높은 문서 5개씩 출력

import random

# Assuming 'similarity_matrix' and 'df' are defined from the previous code

# Select 5 random documents
random_indices = random.sample(range(len(df)), 5)

for i in random_indices:
    # Get similarity scores for the current document
    similarity_scores = similarity_matrix[i]

    # Get the indices of the top 5 most similar documents (excluding itself)
    top_5_indices = np.argsort(similarity_scores)[:-6:-1]  # Get top 5, excluding the document itself

    print(f"\nDocument {i +1}: {df['Related Title'].iloc[i]}")  # Print the selected document's title
    print("Top 5 most similar documents:")
    for j in top_5_indices:
        print(f"  - Document {j +1}: {df['Related Title'].iloc[j]}, Similarity: {similarity_scores[j]:.4f}")
```

```python
# prompt: df['Combined Document'] 를 통해 5개로 군집화 하고 마지막 컬럼에 군집값을 넣어주고 각 군집별로 상위 키워드 20개와 가중치 출력, 각 군집에 할당된 데이터 수도 표시 및 각 군집별 빈도수 시각화

from sklearn.cluster import KMeans

# Assuming 'df' is your DataFrame with the 'Combined Document' column
# ... (your existing code to load and preprocess data) ...

# Prepare the text data for clustering (using TF-IDF)
vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')
tfidf = vectorizer.fit_transform(df['Combined Document'].dropna().astype(str))

# Apply KMeans clustering
num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df['Cluster'] = kmeans.fit_predict(tfidf)

# Print cluster information
for cluster_id in range(num_clusters):
    cluster_data = df[df['Cluster'] == cluster_id]
    print(f"\nCluster {cluster_id + 1}:")
    print(f"  Number of documents: {len(cluster_data)}")

    # Get top keywords for each cluster
    cluster_tfidf = tfidf[df['Cluster'] == cluster_id]
    feature_names = vectorizer.get_feature_names_out()
    top_keywords = []
    for i in range(cluster_tfidf.shape[0]):  # Iterate through documents in the cluster
        top_indices = np.argsort(cluster_tfidf[i].toarray()[0])[:-21:-1]
        top_keywords.extend([(feature_names[j], cluster_tfidf[i, j]) for j in top_indices])

    # Count keyword occurrences and get top 20
    keyword_counts = {}
    for keyword, weight in top_keywords:
        if keyword not in keyword_counts:
            keyword_counts[keyword] = 0
        keyword_counts[keyword] += weight

    sorted_keywords = sorted(keyword_counts.items(), key=lambda x: x[1], reverse=True)[:20]
    print("  Top 20 keywords:")
    for keyword, weight in sorted_keywords:
        print(f"  - {keyword}: {weight}")

# Visualize cluster distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Cluster', data=df)
plt.title('Distribution of Documents across Clusters')
plt.xlabel('Cluster')
plt.ylabel('Number of Documents')
plt.show()
```
