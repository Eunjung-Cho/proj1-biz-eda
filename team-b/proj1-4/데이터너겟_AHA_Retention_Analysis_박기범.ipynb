```python
import json
from pandas import json_normalize

# 1) 문자열 -> 파이썬 dict로 변환
# 2) dict 안의 'device' 키에 해당하는 딕셔너리만 추출
# 3) 그걸 바로 json_normalize로 펼쳐서 데이터프레임화
df_device_flat = (
    df['device']
      .apply(json.loads)              # 문자열 -> dict
      .apply(lambda x: x.get('device', {}))  # "device" 키값 추출
      .pipe(json_normalize)           # dict -> df로 펼치기
)

# device.* 접두사가 붙는 경우가 있다면, 아래처럼 컬럼명 간단히 조정
# df_device_flat.columns = df_device_flat.columns.str.replace('device.', '')

# 4) 기존 df와 좌우로 합치기
df = pd.concat([df, df_device_flat], axis=1)

# 잘 합쳐졌는지 확인
df.head()
```

```python
import pandas as pd
import json
from pandas import json_normalize

# 예시: JSON 형태로 파싱해야 할 컬럼들 목록
json_cols = [
    'event_params', 'device', 'geo', 'app_info', 'traffic_source',
    'user_properties', 'user_ltv', 'event_dimensions', 'ecommerce',
    'items', 'collected_traffic_source', 'session_traffic_source_last_click',
    'publisher', 'privacy_info'
]

def parse_json(x):
    """
    JSON 문자열을 dict로 안전하게 변환.
    파싱 실패 시 빈 dict를 반환.
    """
    try:
        return json.loads(x)
    except:
        return {}

# 1) JSON 파싱 & 펼치기
for col in json_cols:
    # 1) 각 컬럼을 dict로 변환
    parsed_series = df[col].apply(parse_json)

    # 2) dict를 펼쳐서 DataFrame으로 만들기
    df_flat = json_normalize(parsed_series)

    # 3) 중복되지 않도록 새로운 컬럼명에 prefix를 추가
    df_flat.columns = [f"{col}.{subcol}" for subcol in df_flat.columns]

    # 4) 기존 df와 좌우로 합치기
    df = pd.concat([df, df_flat], axis=1)

# 2) 원본 JSON 컬럼 제거
df.drop(columns=json_cols, inplace=True)

# 결과 확인
df.head()
```

```python
# prompt: df.head(1) 을 transpose하고, 모든 header들이 다 표현되도록 할것 (100개가 넘어감). 중간에 ... 으로 생략하지 않도록

import pandas as pd
pd.set_option("display.max_rows", None, "display.max_columns", None)
display(df.head(1).T)
```

아래 코드는 사용자가 plan_start_click 이벤트를 클릭하기 전까지 어떤 이벤트들을 몇 회씩 경험했는지를 분석하는 예제입니다. 이 분석을 통해 AHA Moment 분석 관점에서, 사용자가 어떤 행동들을 여러 번 거친 후 plan_start_click에 도달하는지 파악할 수 있습니다.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 로드 및 전처리 (예시)
# df = pd.read_csv("your_data.csv")
# event_timestamp가 마이크로초 단위라고 가정하여 datetime 변환
df["event_timestamp"] = pd.to_datetime(df["event_timestamp"], unit="us")

# target event 지정
target_event = "plan_start_click"

# 각 사용자가 plan_start_click을 처음 클릭한 시점 구하기
target_times = (
    df[df["event_name"] == target_event]
    .groupby("user_pseudo_id")["event_timestamp"]
    .min()
    .reset_index()
    .rename(columns={"event_timestamp": "target_time"})
)

# plan_start_click을 경험한 사용자만 대상으로 모든 이벤트 병합 (inner join)
merged_df = pd.merge(df, target_times, on="user_pseudo_id", how="inner")

# target event 이전에 발생한 이벤트 필터링
pre_target_df = merged_df[merged_df["event_timestamp"] < merged_df["target_time"]]

# 사용자별로, target event 전 각 이벤트의 발생 횟수를 계산
user_event_counts = (
    pre_target_df.groupby(["user_pseudo_id", "event_name"])
    .size()
    .reset_index(name="event_count")
)

# 모든 사용자를 대상으로 각 이벤트의 평균 발생 횟수 계산 (AHA Moment 전의 행동 패턴)
avg_event_counts = (
    user_event_counts.groupby("event_name")["event_count"]
    .mean()
    .reset_index()
    .sort_values(by="event_count", ascending=False)
)

print("각 이벤트의 평균 발생 횟수 (plan_start_click 전):")
print(avg_event_counts)

# 결과를 바 차트로 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x="event_count", y="event_name", data=avg_event_counts, palette="Blues_d")
plt.title("평균 사전 이벤트 발생 횟수 (plan_start_click 전)")
plt.xlabel("평균 이벤트 횟수")
plt.ylabel("이벤트 이름")
plt.show()
```

plan_start_click 을 클릭한 유저와 그렇지 않은 유저들의 "plan_start_click" 이전 event 발생 회수 비교

```python
import pandas as pd

# 예시: df는 user_pseudo_id, event_name 컬럼을 포함한다고 가정

# 1. plan_start_click 이벤트를 한 사용자의 집합 구하기
users_with_plan = set(df[df["event_name"] == "plan_start_click"]["user_pseudo_id"].unique())

# 2. Group A (plan_start_click이 있는 사용자)와 Group B (없는 사용자)로 데이터 분리
df_A = df[df["user_pseudo_id"].isin(users_with_plan)]
df_B = df[~df["user_pseudo_id"].isin(users_with_plan)]

# 3. 각 그룹별로 사용자별 이벤트 발생 횟수를 계산 (피벗 테이블 형태)
#    - 행: user_pseudo_id, 열: event_name, 값: 해당 이벤트 발생 횟수 (없으면 0)
groupA_counts = df_A.groupby(["user_pseudo_id", "event_name"]).size().unstack(fill_value=0)
groupB_counts = df_B.groupby(["user_pseudo_id", "event_name"]).size().unstack(fill_value=0)

# 4. 각 이벤트별, 그룹 내 전체 사용자를 기준으로 평균 발생 횟수 계산
#    (즉, 이벤트를 아예 수행하지 않은 사용자도 0으로 포함)
avg_counts_A = groupA_counts.mean().rename("avg_count_A").reset_index()
avg_counts_A.rename(columns={"index": "event_name"}, inplace=True)  # index 컬럼명 재설정
avg_counts_B = groupB_counts.mean().rename("avg_count_B").reset_index()
avg_counts_B.rename(columns={"index": "event_name"}, inplace=True)

# 5. 두 그룹의 평균 발생 횟수를 이벤트별로 병합하여 비교
avg_counts = pd.merge(avg_counts_A, avg_counts_B, on="event_name", how="outer").fillna(0)

# 출력

print("\n두 그룹의 이벤트별 평균 발생 횟수 비교:")
print(avg_counts)
```

```python
# prompt: 윗 셀에서  groupA_counts와 groupB_counts의 기술통계

# 기술통계량 계산 및 출력
print("\nGroup A (plan_start_click 있는 사용자) 이벤트 발생 횟수 기술통계:")
print(groupA_counts.describe())

print("\nGroup B (plan_start_click 없는 사용자) 이벤트 발생 횟수 기술통계:")
print(groupB_counts.describe())
```

```python
# prompt: avg_counts 시각화

import matplotlib.pyplot as plt
# 결과를 바 차트로 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x="event_name", y="avg_count_A", data=avg_counts, label="plan_start_click 그룹", color="skyblue")
sns.barplot(x="event_name", y="avg_count_B", data=avg_counts, label="plan_start_click 없는 그룹", color="lightcoral")
plt.title("plan_start_click 여부에 따른 이벤트 발생 횟수 비교")
plt.xlabel("이벤트 이름")
plt.ylabel("평균 이벤트 횟수")
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

Group A와 Group B가 유저수 차이가 많이 나므로 이 부분을 감안해서 통계분석 진행
(Mann–Whitney U 검정 결과)

```python
import pandas as pd
from scipy.stats import mannwhitneyu

# -----------------------------
# 1. 사용자 그룹 분리 및 이벤트 집계
# -----------------------------
# plan_start_click 이벤트를 발생시킨 사용자 ID 집합 (Group A)
users_with_plan = set(df[df["event_name"] == "plan_start_click"]["user_pseudo_id"].unique())

# Group A: plan_start_click 이벤트를 발생시킨 사용자, Group B: 발생시키지 않은 사용자
df_A = df[df["user_pseudo_id"].isin(users_with_plan)]
df_B = df[~df["user_pseudo_id"].isin(users_with_plan)]

# 사용자별, 이벤트별 발생 횟수를 피벗 테이블로 계산 (없으면 0으로 채움)
groupA_counts = df_A.groupby(["user_pseudo_id", "event_name"]).size().unstack(fill_value=0)
groupB_counts = df_B.groupby(["user_pseudo_id", "event_name"]).size().unstack(fill_value=0)

# 모든 이벤트 이름의 목록 생성 (두 그룹 모두에 포함되도록)
all_events = sorted(set(groupA_counts.columns).union(set(groupB_counts.columns)))

# 두 그룹 모두 모든 이벤트 컬럼으로 재정렬 (없으면 0으로 채움)
groupA_counts = groupA_counts.reindex(columns=all_events, fill_value=0)
groupB_counts = groupB_counts.reindex(columns=all_events, fill_value=0)

# -----------------------------
# 2. 각 이벤트별로 Group A와 Group B의 발생 횟수 차이를 Mann–Whitney U 검정으로 비교
# -----------------------------
results = []
for event in all_events:
    counts_A = groupA_counts[event]
    counts_B = groupB_counts[event]

    # Mann–Whitney U 검정 수행 (양측 검정)
    try:
        stat, p_value = mannwhitneyu(counts_A, counts_B, alternative='two-sided')
    except Exception as e:
        stat, p_value = None, None

    results.append({
        'event': event,
        'mean_A': counts_A.mean(),
        'mean_B': counts_B.mean(),
        'statistic': stat,
        'p_value': p_value
    })

results_df = pd.DataFrame(results)

print("각 이벤트별 Group A와 Group B의 발생 횟수 비교 (Mann–Whitney U 검정 결과):")
print(results_df)
```

아래 결과를 보면, **plan_start_click** AHA Moment를 경험한 그룹(A)과 그렇지 않은 그룹(B) 간에 여러 이벤트의 발생 횟수에 상당한 차이가 있음을 알 수 있습니다. 이를 AHA Moment 관점에서 해석하면 다음과 같습니다.

## 결론

**plan_start_click** 이벤트를 발생시키는 AHA Moment는, 사용자가 단순히 첫 방문하는 것을 넘어 **깊은 제품 탐색, 반복적인 콘텐츠 소비, 그리고 계정 연결 및 개인화**를 통해 제품의 핵심 가치를 체감하는 순간이라고 볼 수 있습니다.  
따라서, 제품의 초기 온보딩 전략과 UX/UI 개선에서 이들 핵심 행동을 강조하고 쉽게 접근할 수 있도록 유도하는 것이, 보다 많은 사용자가 AHA Moment를 경험하고 결국 plan_start_click 같은 전환으로 이어지게 하는 중요한 전략이 될 것입니다.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 가정: df는 이미 로드되어 있으며, "user_pseudo_id", "event_timestamp", "event_name" 컬럼이 있음
# event_timestamp는 마이크로초 단위이므로 datetime으로 변환
df["event_timestamp"] = pd.to_datetime(df["event_timestamp"], unit="us")

# Group A: plan_start_click 이벤트가 발생한 사용자
group_A_ids = set(df[df["event_name"] == "plan_start_click"]["user_pseudo_id"].unique())

# Group B: plan_start_click 이벤트가 발생하지 않은 사용자
all_users = set(df["user_pseudo_id"].unique())
group_B_ids = all_users - group_A_ids

# session_start 이벤트만 사용 (리텐션 지표로 활용)
df_session = df[df["event_name"] == "session_start"][["user_pseudo_id", "event_timestamp"]]

def compute_retention(df_session, user_ids):
    """
    주별 리텐션 행렬을 계산하는 함수.
    각 사용자의 첫 session_start 시점을 기준(cohort)으로 하여,
    이후 주차별 unique 사용자 수 비율을 계산합니다.
    """
    # 해당 그룹의 session_start 이벤트 필터링
    group_df = df_session[df_session["user_pseudo_id"].isin(user_ids)].copy()

    # 각 사용자의 첫 session_start를 찾고, cohort_week(주 단위)로 정의
    first_sessions = group_df.groupby("user_pseudo_id")["event_timestamp"].min().reset_index()
    first_sessions["cohort_week"] = first_sessions["event_timestamp"].dt.to_period("W")

    # 그룹 데이터에 cohort_week 병합
    group_df = group_df.merge(first_sessions[["user_pseudo_id", "cohort_week"]], on="user_pseudo_id", how="left")
    group_df["session_week"] = group_df["event_timestamp"].dt.to_period("W")

    # 첫 session_week와 현재 session_week의 차이(주 단위)를 계산
    group_df["week_number"] = (group_df["session_week"] - group_df["cohort_week"]).apply(lambda x: x.n)

    # 각 cohort(첫 session_week)와 이후 주차별로 unique 사용자 수 집계
    retention = group_df.pivot_table(index="cohort_week",
                                     columns="week_number",
                                     values="user_pseudo_id",
                                     aggfunc=pd.Series.nunique)

    # 각 코호트의 첫 주 사용자 수
    cohort_sizes = retention.iloc[:, 0]
    # 리텐션율 계산: 각 주차의 사용자 수를 첫 주 사용자 수로 나눔
    retention_rate = retention.divide(cohort_sizes, axis=0)
    return retention_rate

# Group A와 Group B에 대해 리텐션 행렬 계산
retention_A = compute_retention(df_session, group_A_ids)
retention_B = compute_retention(df_session, group_B_ids)

# 결과 출력
print("Group A (plan_start_click 이벤트 발생 사용자) 리텐션:")
print(retention_A)
print("\nGroup B (plan_start_click 이벤트 미발생 사용자) 리텐션:")
print(retention_B)

# 히트맵 시각화: 두 그룹의 리텐션 행렬을 비교
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.heatmap(retention_A, annot=True, fmt=".0%", cmap="Blues", cbar=False)
plt.title("Group A Retention (plan_start_click 사용자)")
plt.xlabel("Weeks since first session")
plt.ylabel("Cohort Week")

plt.subplot(1, 2, 2)
sns.heatmap(retention_B, annot=True, fmt=".0%", cmap="Greens", cbar=False)
plt.title("Group B Retention (비 plan_start_click 사용자)")
plt.xlabel("Weeks since first session")
plt.ylabel("Cohort Week")

plt.tight_layout()
plt.show()
```

  dashboard_enter_click = 1 을 기준으로 그룹을 나눠서 리텐션 분석 다시 진행

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 가정: df는 이미 로드되어 있으며, "user_pseudo_id", "event_timestamp", "event_name" 컬럼을 포함합니다.
# event_timestamp는 마이크로초 단위이므로 datetime으로 변환합니다.
df["event_timestamp"] = pd.to_datetime(df["event_timestamp"], unit="us")

# 1. 각 사용자의 dashboard_enter_click 이벤트 발생 횟수 계산
dashboard_enter_counts = (
    df[df["event_name"] == "dashboard_enter_click"]
    .groupby("user_pseudo_id")
    .size()
    .reset_index(name="dashboard_enter_count")
)

# 모든 사용자를 포함하도록, 해당 이벤트가 없는 사용자는 0으로 채움
all_users = df["user_pseudo_id"].unique()
dashboard_enter_counts = dashboard_enter_counts.set_index("user_pseudo_id").reindex(all_users, fill_value=0).reset_index()

# 2. Threshold를 dashboard_enter_click = 1 로 설정하여 그룹 분리
threshold = 1
high_group_ids = set(dashboard_enter_counts[dashboard_enter_counts["dashboard_enter_count"] >= threshold]["user_pseudo_id"])
low_group_ids = set(dashboard_enter_counts[dashboard_enter_counts["dashboard_enter_count"] < threshold]["user_pseudo_id"])

print(f"전체 사용자 수: {len(all_users)}")
print(f"High Engagement Group (dashboard_enter_click >= {threshold}): {len(high_group_ids)}")
print(f"Low Engagement Group (dashboard_enter_click < {threshold}): {len(low_group_ids)}")

# 3. 리텐션 분석을 위한 함수 정의 (session_start 이벤트 기반)
def compute_retention(df_session, user_ids):
    """
    각 사용자의 첫 session_start를 기준으로 주별 리텐션 행렬 계산
    """
    group_df = df_session[df_session["user_pseudo_id"].isin(user_ids)].copy()

    # 각 사용자의 첫 session_start 시점을 구해서 코호트로 설정
    first_sessions = group_df.groupby("user_pseudo_id")["event_timestamp"].min().reset_index()
    first_sessions["cohort_week"] = first_sessions["event_timestamp"].dt.to_period("W")

    # 코호트 정보를 원본 데이터에 병합
    group_df = group_df.merge(first_sessions[["user_pseudo_id", "cohort_week"]], on="user_pseudo_id", how="left")
    group_df["session_week"] = group_df["event_timestamp"].dt.to_period("W")

    # 첫 session_week와 현재 session_week 간의 차이(주 단위)
    group_df["week_number"] = (group_df["session_week"] - group_df["cohort_week"]).apply(lambda x: x.n)

    # 코호트별, 주차별로 unique 사용자 수 집계
    retention = group_df.pivot_table(index="cohort_week",
                                     columns="week_number",
                                     values="user_pseudo_id",
                                     aggfunc=pd.Series.nunique)

    # 각 코호트의 첫 주 사용자 수로 나누어 리텐션율 계산
    cohort_sizes = retention.iloc[:, 0]
    retention_rate = retention.divide(cohort_sizes, axis=0)
    return retention_rate

# 4. session_start 이벤트 데이터 필터링
df_session = df[df["event_name"] == "session_start"][["user_pseudo_id", "event_timestamp"]]

# 5. 각 그룹별 리텐션 행렬 계산
retention_high = compute_retention(df_session, high_group_ids)
retention_low = compute_retention(df_session, low_group_ids)

# 6. 리텐션 결과 출력 및 히트맵 시각화
print("High Engagement Group (dashboard_enter_click >= 1) 리텐션:")
print(retention_high)
print("\nLow Engagement Group (dashboard_enter_click < 1) 리텐션:")
print(retention_low)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.heatmap(retention_high, annot=True, fmt=".0%", cmap="Blues", cbar=False)
plt.title("High Engagement Retention (dashboard_enter_click >= 1)")
plt.xlabel("Weeks since first session")
plt.ylabel("Cohort Week")

plt.subplot(1, 2, 2)
sns.heatmap(retention_low, annot=True, fmt=".0%", cmap="Greens", cbar=False)
plt.title("Low Engagement Retention (dashboard_enter_click < 1)")
plt.xlabel("Weeks since first session")
plt.ylabel("Cohort Week")

plt.tight_layout()
plt.show()
```
