```python
import json
import pandas as pd
import numpy as np

def parse_ga4_style_json(json_str):
    """
    1) JSON 문자열인 json_str을 파싱
    2) GA4 'key'/'value' 구조([{ "key": k, "value": {...}}])를 찾으면 -> {k: 실제값}으로 추출
    3) 일반 딕셔너리({ "some_key": some_val }) 구조도 최상위 key-value만 추출
    4) key-value를 담은 dict를 리턴 (실패 시 빈 dict)
    """
    # 예외처리: JSON 파싱
    try:
        parsed = json.loads(json_str)
    except:
        return {}

    # 결과를 담을 딕셔너리
    result = {}

    # 1) 리스트 형태(["event_params": [...]] 등) 안에서 GA4 style param 찾기
    if isinstance(parsed, dict):
        # 예: {"event_params": [{"key": "...", "value": {...}}, ...], "other_key": {...}}
        for top_key, top_val in parsed.items():

            # (A) 만약 리스트라면 → GA4 파라미터(또는 유사 구조) 가능성
            if isinstance(top_val, list):
                for item in top_val:
                    if isinstance(item, dict) and "key" in item and "value" in item:
                        # item: {"key": "...", "value": {...}}
                        param_key = item["key"]
                        param_val = item["value"]

                        # 실제 값 추출 (int_value, string_value, float_value, double_value 중 유효한 것)
                        if param_val.get("int_value") is not None:
                            val = int(param_val["int_value"])
                        elif param_val.get("string_value") is not None:
                            val = param_val["string_value"]
                        elif param_val.get("float_value") is not None:
                            val = float(param_val["float_value"])
                        elif param_val.get("double_value") is not None:
                            val = float(param_val["double_value"])
                        else:
                            val = None

                        result[param_key] = val

            # (B) 딕셔너리 형태라면(예: {"publisher": null}처럼 단일 key-value) → 최상위 키/값을 넣어줌
            elif isinstance(top_val, dict):
                # 간단히 top_val를 평탄화해서 result에 합침 (중첩은 추가 구현 필요)
                for inner_k, inner_v in top_val.items():
                    # key가 "key"이고 value가 {...} 이런 GA4 형식이 아닌 한, 그냥 key-value로 저장
                    result[f"{top_key}_{inner_k}"] = inner_v

            # (C) top_val이 숫자, 문자열, None 등 스칼라일 경우
            else:
                result[top_key] = top_val

    return result


def flatten_ga4_json_columns(df):
    """
    1) df 내에서 '{'로 시작하는 문자열이 있는 컬럼을 'JSON 컬럼'으로 판단
    2) 각 JSON 컬럼을 parse_ga4_style_json으로 변환 -> 여러 컬럼으로 펼침
    3) 펼쳐진 컬럼을 원본 df와 병합 후, 기존 JSON 컬럼은 삭제
    4) 최종 df 반환
    """
    json_cols = []

    # 1) 우선, '{'로 시작하는 문자열이 있는 컬럼 식별
    for col in df.columns:
        # Series를 문자열화한 뒤 '{'로 시작하는지(any) 확인
        if df[col].astype(str).str.startswith('{').any():
            json_cols.append(col)

    # 2) 각 JSON 컬럼별로 파싱 & 펼치기
    expanded_dfs = []
    for col in json_cols:
        # parse 결과 -> dict 형태가 row별로 생김
        parsed_series = df[col].astype(str).apply(parse_ga4_style_json)

        # dict -> 데이터프레임으로 (row별 컬럼 확장)
        expanded_df = pd.json_normalize(parsed_series)

        # 혹시 컬럼 충돌 방지를 위해 '원본컬럼명.새키' 형태로 명명
        expanded_df.columns = [f"{col}.{c}" for c in expanded_df.columns]

        expanded_dfs.append(expanded_df)

    # 3) 확장된 DF를 모두 수평 결합(concat) -> 원본 DF와 합치기
    if expanded_dfs:
        expanded_all = pd.concat([df] + expanded_dfs, axis=1)
    else:
        expanded_all = df.copy()

    # 4) 기존 JSON 컬럼은 삭제
    expanded_all.drop(columns=json_cols, inplace=True)

    return expanded_all


# 이미 df가 존재한다고 가정; 여기서는 예시로 일부 컬럼만 만듦
# 변환 수행
df_flattened = flatten_ga4_json_columns(df)
print("=== Before ===")
print(df)
print("\n=== After Flattening ===")
print(df_flattened)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 기본 스타일 설정
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['axes.grid'] = True
plt.rcParams['font.size'] = 10

# 데이터 로드
df = pd.read_csv('df_flattened_final.csv')

# 1. 이벤트 분포 분석
plt.figure(figsize=(12, 6))
top_events = df['event_name'].value_counts().head(10)
ax = sns.barplot(x=top_events.values, y=top_events.index, color='skyblue')
plt.title('Top 10 Most Frequent Events')
plt.xlabel('Count')
plt.ylabel('Event Name')

# 값 레이블 추가
for i, v in enumerate(top_events.values):
    ax.text(v, i, f' {v:,}', va='center')

plt.tight_layout()
plt.savefig('top_events.png')
plt.close()

# 2. 플랫폼별 이벤트 분포
plt.figure(figsize=(10, 6))
platform_dist = df['platform'].value_counts()
plt.pie(platform_dist.values, labels=platform_dist.index,
        autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Platform Distribution')
plt.axis('equal')
plt.savefig('platform_distribution.png')
plt.close()

# 3. 이벤트-플랫폼 관계 분석
plt.figure(figsize=(12, 8))
event_platform = pd.crosstab(df['event_name'], df['platform'])
event_platform_pct = event_platform.div(event_platform.sum(axis=1), axis=0)

# 상위 10개 이벤트만 선택
top_10_events = df['event_name'].value_counts().head(10).index
event_platform_pct = event_platform_pct.loc[top_10_events]

sns.heatmap(event_platform_pct, annot=True, fmt='.2%', cmap='YlOrRd')
plt.title('Event Distribution by Platform (Top 10 Events)')
plt.ylabel('Event Name')
plt.xlabel('Platform')
plt.tight_layout()
plt.savefig('event_platform_heatmap.png')
plt.close()

# 4. 사용자별 이벤트 분석
user_analysis = df.groupby('user_pseudo_id').agg({
    'event_name': ['count', 'nunique']
}).reset_index()
user_analysis.columns = ['user_id', 'total_events', 'unique_events']

plt.figure(figsize=(10, 6))
plt.scatter(user_analysis['total_events'], user_analysis['unique_events'],
           alpha=0.5, color='blue')
plt.title('User Event Patterns')
plt.xlabel('Total Number of Events')
plt.ylabel('Number of Unique Events')
plt.tight_layout()
plt.savefig('user_event_patterns.png')
plt.close()

# 5. 이벤트 시퀀스 분석
df_sorted = df.sort_values(['user_pseudo_id', 'event_timestamp'])
df_sorted['next_event'] = df_sorted.groupby('user_pseudo_id')['event_name'].shift(-1)
event_pairs = df_sorted.groupby(['event_name', 'next_event']).size()

# 상위 10개 이벤트 시퀀스 시각화
plt.figure(figsize=(12, 6))
top_pairs = event_pairs.nlargest(10)
sns.barplot(x=top_pairs.values, y=[f"{a} → {b}" for a, b in top_pairs.index],
            color='lightgreen')
plt.title('Top 10 Event Sequences')
plt.xlabel('Frequency')
plt.ylabel('Event Sequence')
plt.tight_layout()
plt.savefig('event_sequences.png')
plt.close()

# 기본 통계 출력
print("=== 기본 통계 ===")
print(f"총 이벤트 수: {len(df):,}")
print(f"유니크 사용자 수: {df['user_pseudo_id'].nunique():,}")
print(f"유니크 이벤트 타입: {df['event_name'].nunique():,}")

print("\n=== 상위 5개 이벤트 ===")
print(df['event_name'].value_counts().head())

print("\n=== 플랫폼별 통계 ===")
platform_stats = df.groupby('platform').agg({
    'event_name': 'count',
    'user_pseudo_id': 'nunique'
}).rename(columns={
    'event_name': '이벤트 수',
    'user_pseudo_id': '유니크 사용자 수'
})
print(platform_stats)

# 사용자 세그먼트 분석
user_segments = pd.qcut(user_analysis['total_events'], q=4, labels=['Low', 'Medium', 'High', 'Very High'])
segment_stats = pd.DataFrame({
    'Segment': user_segments.value_counts().index,
    'Users': user_segments.value_counts().values,
    'Percentage': user_segments.value_counts(normalize=True).values * 100
})

print("\n=== 사용자 세그먼트 분석 ===")
print(segment_stats)

# engagement_time 분석 (컬럼이 있는 경우)
if 'event_params.engagement_time_msec' in df.columns:
    engagement_stats = df[df['event_params.engagement_time_msec'].notna()].groupby('event_name')['event_params.engagement_time_msec'].agg([
        'mean', 'median', 'count'
    ]).sort_values('mean', ascending=False)

    print("\n=== 이벤트별 평균 체류시간 (초) ===")
    engagement_stats['mean_seconds'] = engagement_stats['mean'] / 1000
    engagement_stats['median_seconds'] = engagement_stats['median'] / 1000
    print(engagement_stats.head(10))

# 이벤트 연관성 분석 결과 출력
print("\n=== 가장 빈번한 이벤트 시퀀스 TOP 10 ===")
for (event1, event2), count in event_pairs.nlargest(10).items():
    print(f"{event1} → {event2}: {count:,}회")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 기본 스타일 설정
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['axes.grid'] = True
plt.rcParams['font.size'] = 10

# 데이터 로드
df = pd.read_csv('df_flattened_final.csv')

# timestamp를 단순히 24시간으로 나누어 처리
df['hour'] = (df['event_timestamp'] % (24 * 3600 * 1000)) // (3600 * 1000)

# 1. 시간대별 이벤트 분포 (Heatmap)
plt.figure(figsize=(15, 8))
top_events = df['event_name'].value_counts().head(10).index
hourly_events = pd.crosstab(df['hour'], df['event_name'][df['event_name'].isin(top_events)])

sns.heatmap(hourly_events, cmap='YlOrRd', annot=True, fmt='g')
plt.title('Hourly Distribution of Top 10 Events')
plt.xlabel('Event Type')
plt.ylabel('Hour of Day')
plt.tight_layout()
plt.savefig('hourly_events_heatmap.png')
plt.close()

# 2. 시간대별 플랫폼 사용 패턴
plt.figure(figsize=(12, 6))
platform_hourly = pd.crosstab(df['hour'], df['platform'])
platform_hourly.plot(kind='line', marker='o')
plt.title('Platform Usage by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Events')
plt.legend(title='Platform')
plt.grid(True)
plt.tight_layout()
plt.savefig('platform_hourly_usage.png')
plt.close()

# 3. 시간대별 총 이벤트 수
plt.figure(figsize=(12, 6))
hourly_total = df.groupby('hour').size()
sns.lineplot(x=hourly_total.index, y=hourly_total.values, marker='o')
plt.title('Total Events by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Events')
plt.grid(True)
plt.tight_layout()
plt.savefig('hourly_total_events.png')
plt.close()

# 4. 시간대별 평균 체류시간
if 'event_params.engagement_time_msec' in df.columns:
    plt.figure(figsize=(12, 6))
    hourly_engagement = df[df['event_params.engagement_time_msec'].notna()].groupby('hour')['event_params.engagement_time_msec'].mean() / 1000
    sns.lineplot(x=hourly_engagement.index, y=hourly_engagement.values, marker='o')
    plt.title('Average Engagement Time by Hour')
    plt.xlabel('Hour of Day')
    plt.ylabel('Average Engagement Time (seconds)')
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('hourly_engagement.png')
    plt.close()

# 5. 시간대별 유니크 사용자 수
plt.figure(figsize=(12, 6))
hourly_users = df.groupby('hour')['user_pseudo_id'].nunique()
sns.lineplot(x=hourly_users.index, y=hourly_users.values, marker='o')
plt.title('Unique Users by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Unique Users')
plt.grid(True)
plt.tight_layout()
plt.savefig('hourly_users.png')
plt.close()

# 통계 출력
print("=== 시간대별 분석 ===")
hourly_stats = pd.DataFrame({
    '총 이벤트 수': hourly_total,
    '유니크 사용자 수': hourly_users,
    '평균 이벤트/사용자': hourly_total / hourly_users
})
print("\n시간대별 기본 통계:")
print(hourly_stats.sort_values('총 이벤트 수', ascending=False))

# 피크 시간대 분석
peak_hour = hourly_total.idxmax()
print(f"\n피크 시간대: {peak_hour}시")
print(f"피크 시간 이벤트 수: {hourly_total[peak_hour]:,}건")
print(f"피크 시간 유니크 사용자: {hourly_users[peak_hour]:,}명")

# 플랫폼별 시간대 통계
print("\n=== 플랫폼별 시간대 통계 ===")
platform_hourly_stats = pd.crosstab(df['hour'], df['platform'])
platform_hourly_pct = platform_hourly_stats.div(platform_hourly_stats.sum(axis=1), axis=0)
print("\n시간대별 플랫폼 비중 (%):")
print((platform_hourly_pct * 100).round(1))

# 시간대별 주요 이벤트
print("\n=== 시간대별 Top 3 이벤트 ===")
for hour in sorted(df['hour'].unique()):
    hour_events = df[df['hour'] == hour]['event_name'].value_counts().head(3)
    if len(hour_events) > 0:
        print(f"\n{hour}시:")
        for event, count in hour_events.items():
            print(f"- {event}: {count}건")

# 플랫폼별 피크타임 분석
ios_peak = platform_hourly_stats['IOS'].idxmax()
android_peak = platform_hourly_stats['ANDROID'].idxmax()
print(f"\n=== 플랫폼별 피크타임 ===")
print(f"iOS 피크타임: {ios_peak}시 ({platform_hourly_stats.loc[ios_peak, 'IOS']}건)")
print(f"Android 피크타임: {android_peak}시 ({platform_hourly_stats.loc[android_peak, 'ANDROID']}건)")
```
