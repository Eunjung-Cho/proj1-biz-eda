해당 그룹 내에서 다음 세 가지 **광고 성과 변수**가 일치하는지 확인.

### 📊 **분석 방법 요약**
1. **그룹화 기준:**  
   - 광고_채널  
   - 광고_형식  
   - 광고_타겟팅  
   - 광고_시작일  
   - 광고_종료일  

2. **검증 대상 성과 지표:**  
   - **광고_노출_횟수**  
   - **광고_클릭_횟수**  
   - **광고_지출액**  

3. **일치 여부 확인:**  
   - 그룹별 **최소값과 최대값**이 같은지 확인.  
   - 그룹 내 **표준편차(standard deviation)**가 0이면 완전히 동일한 것.  

---

### ✅ **해석 방법**
1. **일치할 경우:**  
   - 동일한 캠페인에서 각 사용자에게 개별 노출된 결과라고 해석.  
2. **불일치할 경우:**  
   - 사용자별 반응(클릭, 지출)이 다르거나, 실제로 다른 캠페인일 가능성.




```python
# 광고 캠페인 기준으로 그룹화하고, 광고 성과 변수의 일치 여부 분석
# 그룹화 기준: 광고_채널, 광고_형식, 광고_타겟팅, 광고_시작일, 광고_종료일
campaign_group = ad_data.groupby(['광고_채널', '광고_형식', '광고_타겟팅', '광고_시작일', '광고_종료일'])

# 성과 지표(광고_노출_횟수, 광고_클릭_횟수, 광고_지출액)에 대해 최소값과 최대값을 비교하여 일치 여부 확인
result = campaign_group.agg({
    '광고_노출_횟수': ['min', 'max', 'std'],
    '광고_클릭_횟수': ['min', 'max', 'std'],
    '광고_지출액': ['min', 'max', 'std']
})

# 일치 여부를 쉽게 확인할 수 있도록 추가 컬럼 생성
result['노출_일치'] = result[('광고_노출_횟수', 'min')] == result[('광고_노출_횟수', 'max')]
result['클릭_일치'] = result[('광고_클릭_횟수', 'min')] == result[('광고_클릭_횟수', 'max')]
result['지출_일치'] = result[('광고_지출액', 'min')] == result[('광고_지출액', 'max')]

# 일치 여부와 표준편차를 함께 확인

# 최종 결과 반환
result[['노출_일치', '클릭_일치', '지출_일치']].value_counts()
```

### 📊 **캠페인 성과 일치 분석 결과 요약**

1. **완전히 일치한 캠페인:**  
   - **95,552건 (95.6%)**의 그룹에서 **광고_노출_횟수**, **광고_클릭_횟수**, **광고_지출액**이 모두 일치했습니다.  
   - 이는 대부분의 데이터가 **동일 캠페인 내에서 개별 사용자 반응을 기록**한 것임을 의미합니다.

2. **불일치한 캠페인:**  
   - **2,162건 (2.2%)**에서 **노출, 클릭, 지출** 모두 불일치.  
   - **34건 (0.03%)**에서는 클릭만 일치하고, 지출과 노출이 불일치.  
   - **8건 (0.01%)**에서는 노출만 일치하고, 클릭과 지출이 불일치.  

---

### 🔍 **해석 및 결론**
1. **일치 그룹:**  
   - 대부분의 광고 캠페인은 **광고 속성 기준으로 동일한 캠페인**이라고 간주할 수 있습니다.  
   - 개별 사용자가 같은 광고 캠페인에 노출되고, 성과가 기록된 형태입니다.  

2. **불일치 그룹:**  
   - 성과가 불일치하는 캠페인은 **개별 사용자별 광고 반응 차이**나 **광고 비용 산정 방식의 차이**일 가능성이 높습니다.  
   - 광고 집행 기간 동안 성과가 동적으로 변화했을 수도 있습니다.  



```python
# "광고 성과가 모두 일치(True)" 하는 그룹만 추출
matched_result = result[
    (result['노출_일치']) &
    (result['클릭_일치']) &
    (result['지출_일치'])
]
```

```python

# 1) matched_result.index를 DataFrame으로 변환
#    => (광고_채널, 광고_형식, 광고_타겟팅, 광고_시작일, 광고_종료일) 등 그룹화 기준이 컬럼 형태로 들어옴
matched_result_index_df = matched_result.index.to_frame(index=False).drop_duplicates()

# 2) 이제 matched_result_index_df는
#    "광고_채널", "광고_형식", "광고_타겟팅", "광고_시작일", "광고_종료일" 같은 단일 레벨 컬럼만 가짐

# 3) 원본 데이터 ad_data와 조인
matched_data = pd.merge(
    ad_data,
    matched_result_index_df,
    on=['광고_채널', '광고_형식', '광고_타겟팅', '광고_시작일', '광고_종료일'],
    how='inner'
)
```

#### **1. 성과 지표 (Performance Metrics)**
   - **광고 클릭률 (CTR)**: `광고_클릭_횟수 / 광고_노출_횟수 * 100`
     - 광고가 노출된 대비 실제로 클릭된 비율. 높은 CTR은 광고의 관심도를 의미.
   - **전환율 (CVR)**: `전환율` (이미 데이터에 존재)
     - 클릭 후 실제 구매나 원하는 액션이 일어난 비율. 성과의 가장 중요한 지표.
   - **매출액 (Revenue)**: `매출액` (이미 데이터에 존재)
     - 광고로 인해 발생한 실제 수익.
   - **광고 회상률 (Ad Recall Rate)**: `광고_회상률` (이미 데이터에 존재)
     - 광고 노출 후 브랜드를 기억하는 비율.
   - **광고 태도 변화 (Ad Attitude Shift)**: `광고_태도_변화` (이미 데이터에 존재)
     - 광고 노출 전후의 브랜드나 제품에 대한 인식 변화.

---

#### **2. 효율성 지표 (Efficiency Metrics)**
   - **ROAS (광고 수익률)**: `매출액 / 광고_지출액`
     - 광고비 1원당 발생한 수익. ROAS가 1 이상이면 비용 대비 수익이 긍정적.
   - **CPA (전환당 비용)**: `광고_지출액 / (전환율 * 웹사이트_방문_수)`
     - 실제 구매 1건당 소요된 광고비. 낮을수록 효율적.
   - **CPC (클릭당 비용)**: `광고_지출액 / 광고_클릭_횟수`
     - 클릭 1회당 발생한 비용. 광고 집행 비용 관리에 중요.
   - **CPM (노출 1,000회당 비용)**: `광고_지출액 / 광고_노출_횟수 * 1000`
     - 광고 노출 기준 비용. 노출 중심의 광고에서는 CPM을 낮추는 것이 중요.

---

#### **3. 사용자 반응 지표 (User Engagement Metrics)**
   - **웹사이트 방문율 (Visit Rate)**: `웹사이트_방문_수 / 광고_클릭_횟수 * 100`
     - 광고 클릭 후 실제 웹사이트로 유입된 비율.
   - **페이지뷰당 체류율 (Page Views per Visit)**: `페이지뷰 / 웹사이트_방문_수`
     - 방문당 평균 페이지 조회 수. 높은 값은 콘텐츠에 대한 관심도를 의미.
   - **구매 이력과 전환율 관계:** `구매_이력`과 `전환율`의 상관관계 분석.
     - 기존 구매자와 신규 사용자의 전환율 비교.

---

#### **4. 비용 최적화 지표 (Cost Optimization Metrics)**
   - **광고 비용 비율:** `광고_지출액 / 매출액 * 100`
     - 수익 대비 광고비 비율. 30% 이하가 이상적.
   - **채널별 CPA:** 채널별 전환당 비용 비교.
   - **광고 형식별 ROAS:** 형식별 수익률 비교.

---


```python
# ROAS, CTR, CPA 계산
matched_data['CTR'] = (matched_data['광고_클릭_횟수'] / matched_data['광고_노출_횟수']) * 100
matched_data['ROAS'] = matched_data['매출액'] / matched_data['광고_지출액']
matched_data['CPA'] = matched_data['광고_지출액'] / (matched_data['전환율'] * matched_data['웹사이트_방문_수'])

# 캠페인 기준 성과 분석
campaign_performance = matched_data.groupby(['광고_채널', '광고_형식', '광고_타겟팅', '광고_시작일', '광고_종료일']).agg({
    '광고_노출_횟수': 'sum',
    '광고_클릭_횟수': 'sum',
    '광고_지출액': 'sum',
    '매출액': 'sum',
    'CTR': 'mean',
    'ROAS': 'mean',
    'CPA': 'mean'
}).reset_index()

# 결과 테이블 표시
campaign_performance
```
