```python
import pandas as pd
import json

# 재귀적으로 딕셔너리를 평탄하게 만드는 함수
def flatten_dict(d, parent_key='', sep='_'):
    items = {}
    for k, v in d.items():
        new_key = f"{parent_key}{sep}{k}" if parent_key else k
        if isinstance(v, dict):
            items.update(flatten_dict(v, new_key, sep=sep))
        else:
            items[new_key] = v
    return items

# JSON 문자열을 파싱하여 내부 구조에 따라 평탄한 딕셔너리로 변환하는 함수
def parse_json_field(json_str):
    try:
        obj = json.loads(json_str)
    except Exception:
        return {}

    result = {}
    # 최상위가 딕셔너리인 경우 각 key마다 처리
    for k, v in obj.items():
        # 만약 v가 리스트이고, 리스트의 각 원소가 {'key': ..., 'value': ...} 형태라면 처리
        if isinstance(v, list) and v and all(isinstance(item, dict) and 'key' in item and 'value' in item for item in v):
            for item in v:
                param_key = item.get('key')
                value_info = item.get('value', {})
                # 우선순위: string_value > int_value > float_value > double_value
                param_value = value_info.get('string_value')
                if param_value is None:
                    param_value = value_info.get('int_value')
                if param_value is None:
                    param_value = value_info.get('float_value')
                if param_value is None:
                    param_value = value_info.get('double_value')
                result[param_key] = param_value
        # 만약 v가 딕셔너리라면 재귀적으로 평탄화
        elif isinstance(v, dict):
            flattened = flatten_dict(v, parent_key=k)
            result.update(flattened)
        else:
            result[k] = v
    return result

# DataFrame의 각 컬럼을 순회하면서 JSON 문자열을 포함하는 컬럼을 찾아서 파싱 후 확장
for col in list(df.columns):  # 컬럼명을 리스트로 복사하여 순회 (반복 중 컬럼 삭제 방지)
    # pd.api.types.is_string_dtype()를 사용하여 문자열 타입인지 확인
    if pd.api.types.is_string_dtype(df[col]) and not df[col].dropna().empty:
        sample_val = df[col].dropna().iloc[0]
        if isinstance(sample_val, str) and sample_val.strip().startswith('{'):
            # 각 셀마다 JSON 파싱 (결과는 딕셔너리)
            parsed_series = df[col].apply(lambda x: parse_json_field(x) if pd.notnull(x) else {})
            # pd.json_normalize를 사용하여 Series(딕셔너리)로부터 DataFrame 생성
            new_cols = pd.json_normalize(parsed_series)
            # 원본 DataFrame과 병합
            df = pd.concat([df, new_cols], axis=1)
            # 기존 JSON 문자열 컬럼 삭제
            df.drop(columns=[col], inplace=True)

# 최종 DataFrame 확인
print(df.head())
```
