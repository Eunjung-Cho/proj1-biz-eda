```python
# prompt: <class 'pandas.core.frame.DataFrame'>
# RangeIndex: 12 entries, 0 to 11
# Columns: 120 entries, 지역별(1) to 2024.11
# dtypes: object(120)
# memory usage: 11.4+ KB
# 데이터에 대한 기본 기술 통계와 데이터 분석 코드를 한국어 주석과 함께 작성

# pandas, numpy, matplotlib, plotly 라이브러리를 사용하여 데이터 분석을 수행합니다.
import pandas as pd
import numpy as np
import koreanize_matplotlib
import matplotlib.pyplot as plt
import plotly.express as px

# koreanize-matplotlib 라이브러리를 업데이트합니다. (필요시)
# !pip install -Uq koreanize-matplotlib

# Excel 파일을 pandas DataFrame으로 읽어옵니다.
df = pd.read_excel("inbound_recent10y.xlsx")

# 데이터의 크기(행, 열)를 확인합니다.
print(df.shape)

# 데이터의 상위 5개 행을 출력하여 데이터 형태를 확인합니다.
print(df.head())

# 데이터의 정보(데이터 타입, 결측치 등)를 출력합니다.
print(df.info())


# 기술 통계량 계산 (숫자형 열에 대해서만 계산)
# 숫자형 열만 선택
numeric_cols = df.select_dtypes(include=np.number).columns

# 기술 통계량 출력
if not numeric_cols.empty:
    print("\n숫자형 열의 기술 통계량:")
    print(df[numeric_cols].describe())
else:
    print("\n데이터프레임에 숫자형 열이 없습니다.")

# 각 열의 고유값 개수 확인 (범주형 변수 분석에 유용)
print("\n각 열의 고유값 개수:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()}개")


# 결측치 확인
print("\n결측치 개수:")
print(df.isnull().sum())

# 히스토그램을 이용한 데이터 분포 확인 (숫자형 열에 대해서만)
if not numeric_cols.empty:
    df[numeric_cols].hist(figsize=(10, 10))
    plt.suptitle("각 열의 히스토그램", fontsize=16)
    plt.show()

# plotly를 사용한 시각화 (예시: 특정 열에 대한 막대 그래프)

# 시각화할 열 선택 (데이터에 맞게 변경)
if '지역별(1)' in df.columns: # '지역별(1)' 열이 존재하는지 확인
    fig = px.bar(df, x='지역별(1)', y='2024.11', title='지역별 2024년 11월 데이터')
    fig.show()
else:
  print("'지역별(1)' 열이 존재하지 않습니다. 다른 열을 선택하세요.")


# 추가 분석 (필요에 따라 다른 분석 기법 적용)
# 예: 상관관계 분석, 그룹별 통계량 비교 등
```
